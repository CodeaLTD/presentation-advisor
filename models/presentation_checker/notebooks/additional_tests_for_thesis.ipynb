```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Example: list of image file paths
images = [
    "../data/ppt/bullets1/has_bullets/32.png",
    "../data/ppt/graphics/has/Screenshot 2021-10-03 161129.png",
    "../data/ppt/boring_interesting/interesting_slides/136.png",
    "../data/ppt/agenda/yes/13481df1d87666cb43e8936a34f3a599.png",
    "../data/ppt/infographics/with/gent1 (23).png",
    "../data/ppt/pictures/with/Screenshot 2021-10-03 184623.png",
     "../data/ppt/positioning/good/Screenshot 2021-10-03 185255.png",
    "../data/ppt/readable/yes/Screenshot 2021-10-03 160024.png",
    "../data/ppt/size_of_text/ok/50139-corporate marketing strategy ppt.png",
    "../data/ppt/tables/have/4.png",
    # ... up to 10 images
]
labels = ["bullets", "graphics", "interesting", "agenda", "infographics",
          "pictures", "positioning", "readable", "size_of_text", "tables"]

# Create 2x5 grid
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.flatten()

for i, ax in enumerate(axes):
    img = mpimg.imread(images[i])  # read image from path
    ax.imshow(img)
    ax.set_title(labels[i])
    ax.axis('off')

plt.tight_layout()
plt.show()
```
