```python
import random
import pandas as pd

def generate_user_profiles(num_users, types, locations, preferences_options):
    """
    Generate user profiles with specified preferences.

    Parameters:
        num_users (int): Number of user profiles to generate.
        types (list): List of user types.
        locations (list): List of locations.
        preferences_options (list): List of preference options.

    Returns:
        pd.DataFrame: DataFrame containing the generated user profiles.
    """
    user_profiles = []
    for user_id in range(1, num_users + 1):
        user_type = random.choice(types)
        location = random.choice(locations)
        preferences = [
          random.choice(preferences_options) for _ in range(random.randint(1, 4))  # Randomly decide the number of preferences (1 to 4)
        ]
        user_profiles.append({
            'user_id': user_id,
            'type': user_type,
            'location': location,
            'preferences': preferences,
            'preferred_presentation_type': random.choice(['formal', 'creative', 'students', 'kids'])
        })
    return pd.DataFrame(user_profiles)
```

```python
import random
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def generate_presentations_for_users(user_profiles, problems, max_issues_per_presentation=6):
    """
    Generate presentations for a given number of users with random issues marked.

    Parameters:
        user_profiles: The user profiles of all users for which we will generate presentations for.
        problems (list): List of possible problems to mark in presentations.
        max_issues_per_presentation (int): Maximum number of issues per presentation (default: 6).

    Returns:
        pd.DataFrame: A DataFrame containing all generated presentations.
    """

    def generate_presentation(user_id, presentation_count):
        """
        Generate a random presentation with issues marked for a specific user.

        Parameters:
            user_id (int): The ID of the user.
            presentation_count (int): Counter to ensure unique presentation IDs.

        Returns:
            dict: A dictionary representing a presentation with marked issues.
        """
        # Initialize all problems with 0
        presentation = {problem: 0 for problem in problems}
        # Select a random number of problems to mark as 1
        num_problems = random.randint(1, max_issues_per_presentation)
        selected_problems = random.sample(problems, num_problems)
        for problem in selected_problems:
            presentation[problem] = 1
        presentation['user_id'] = user_id
        presentation['presentation_id'] = f"U{user_id}_P{presentation_count}"
        presentation['submission_date'] = generate_random_date()  # Add random submission date
        return presentation

    presentations = []
    presentation_count = 1  # Initialize a counter for unique presentation IDs
    for user_id in user_profiles['user_id']:
        # Each user gets at least one presentation
        presentations.append(generate_presentation(user_id, presentation_count))
        presentation_count += 1
        # Some users randomly get an additional presentation
        if random.choice([True, False]):
            presentations.append(generate_presentation(user_id, presentation_count))
            presentation_count += 1

    return pd.DataFrame(presentations)

```

```python
import numpy as np
import pandas as pd
import random

def generate_article_ratings_with_preferences_and_popularity(
    user_profiles, recommendations, presentations_df, existing_ratings=None, min_rating=1, max_rating=5
):
    """
    Generate article ratings based on user-linked presentations, issues, audience type,
    preferred presentation type, and article popularity. Each rating includes a timestamp.
    
    Parameters:
        user_profiles (pd.DataFrame): DataFrame containing user profiles.
        recommendations (pd.DataFrame): DataFrame containing article recommendations.
        presentations_df (pd.DataFrame): DataFrame linking users to presentations with issues and submission dates.
        existing_ratings (pd.DataFrame, optional): DataFrame containing existing ratings for articles. Defaults to None.
        min_rating (int): Minimum possible rating value.
        max_rating (int): Maximum possible rating value.
    
    Returns:
        pd.DataFrame: A DataFrame containing user_id, article_id, rating, and timestamp.
    """
    ratings = []
    
    # Convert submission_date to datetime for prioritization
    presentations_df["submission_date"] = pd.to_datetime(presentations_df["submission_date"])
    
    # Calculate average article ratings if existing ratings are provided
    if existing_ratings is not None:
        article_avg_ratings = existing_ratings.groupby("article_id")["rating"].mean()
    else:
        article_avg_ratings = pd.Series(dtype=float)
    
    start_date = pd.to_datetime("2005-12-01 00:00:00")
    end_date = pd.to_datetime("now")
    
    for _, user in user_profiles.iterrows():
        user_id = user["user_id"]
        user_preferences = user["preferences"]
        user_type = user["type"]
        preferred_audience = audience_mapping.get(user_type, "general")
        preferred_presentation_type = user["preferred_presentation_type"]

        # Filter presentations linked to this user
        user_presentations = presentations_df[presentations_df["user_id"] == user_id]
        
        # Weight presentations based on recency
        user_presentations["submission_date"] = pd.to_datetime(user_presentations["submission_date"], utc=True)
        current_time = pd.Timestamp.now(tz="UTC")
        user_presentations["recency_weight"] = (
            1 / (1 + (current_time - user_presentations["submission_date"]).dt.days)
        )

        # Aggregate issues for this user's presentations, weighted by recency
        weighted_issues = (
            user_presentations.iloc[:, :-3]
            .multiply(user_presentations["recency_weight"], axis=0)
            .sum()
        )
        
        for _, rec in recommendations.iterrows():
            article_id = rec["article_id"]
            presentation_type = rec["presentation_type"]
            main_issue = rec["Main issue"]
            audience_type = rec["audience_type"]
            
            # Base rating
            rating = (max_rating + min_rating) // 2  # Start with midpoint rating
            
            # Rule 1: Match user preferences with weighted issues
            matching_issues = set(user_preferences) & set(weighted_issues.index[weighted_issues > 0])
            if matching_issues:
                rating += len(matching_issues)  # Increase rating for each matching issue
            
            # Rule 2: Match preferred presentation type
            if preferred_presentation_type == presentation_type:
                rating += 1  # Boost rating for matching presentation type
            else:
                rating -= 0.5  # Penalize for non-matching presentation type
            
            # Rule 3: Match audience type (skip if preferred audience type is "general")
            if preferred_audience != "general":
                if audience_type == preferred_audience:
                    rating += 0.5  # Boost rating for matching audience type
                else:
                    rating -= 0.5  # Penalize for non-matching audience type
            
            # Rule 4: Incorporate main issue
            if main_issue in user_preferences:
                rating += 0.5
            
            # Rule 5: Adjust based on article popularity (average rating)
            if article_id in article_avg_ratings:
                avg_rating = article_avg_ratings[article_id]
                if avg_rating > (max_rating + min_rating) / 2:
                    rating += 0.5  # Boost for higher average ratings
                else:
                    rating -= 0.5  # Penalize for lower average ratings
            
            # Ensure rating is within bounds
            rating = max(min_rating, min(rating, max_rating))
            
            # Generate a random timestamp within the range
            timestamp = np.random.choice(pd.date_range(start=start_date, end=end_date, freq="H"))
            
            # Append the rating
            ratings.append({"user_id": user_id, "article_id": article_id, "rating": rating, "timestamp": timestamp})
    
    return pd.DataFrame(ratings)

def generate_ratings_matrix(user_profiles_df, recommendations, presentations_df, existing_ratings=None):
    """
    Generate the ratings matrix with timestamps.
    """
    return generate_article_ratings_with_preferences_and_popularity(
        user_profiles_df, recommendations, presentations_df, existing_ratings
    )
```

```python
def preprocess_user_profile(user_profile, preferences_options, types, presentation_type_options):
    """
    Preprocess user profile by flattening preferences and binary encoding categorical features.

    Parameters:
        user_profile (pd.DataFrame): DataFrame containing user-specific preferences and attributes.
        preferences_options (list): List of all possible preferences.
        types (list): List of possible user types.
        presentation_type_options (list): List of possible presentation types.

    Returns:
        pd.DataFrame: Updated user_profile with flattened preferences and binary-encoded features.
    """
    # Step 1: Flatten preferences
    preference_columns = {pref: f"user_preference_{pref}" for pref in preferences_options}
    preferences_df = pd.DataFrame(0, index=user_profile.index, columns=preference_columns.values())
    for i, prefs in enumerate(user_profile['preferences']):
        for pref in prefs:
            if pref in preference_columns:
                preferences_df.loc[i, preference_columns[pref]] = 1

    # Step 2: Binary encode 'type'
    type_columns = {t: f"user_type_{t}" for t in types}
    type_df = pd.DataFrame(0, index=user_profile.index, columns=type_columns.values())
    for i, user_type in enumerate(user_profile['type']):
        if user_type in type_columns:
            type_df.loc[i, type_columns[user_type]] = 1

    # Binary encode 'preferred_presentation_type'
    presentation_columns = {pt: f"presentation_type_{pt}" for pt in presentation_type_options}
    presentation_df = pd.DataFrame(0, index=user_profile.index, columns=presentation_columns.values())
    for i, presentation_type in enumerate(user_profile['preferred_presentation_type']):
        if presentation_type in presentation_columns:
            presentation_df.loc[i, presentation_columns[presentation_type]] = 1

    # Concatenate all features with the original DataFrame
    user_profile_encoded = pd.concat([
        user_profile.drop(columns=['preferences', 'type', 'preferred_presentation_type']),
        preferences_df, type_df, presentation_df
    ], axis=1)

    return user_profile_encoded
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

import pandas as pd

import numpy as np
import pandas as pd

import numpy as np
import pandas as pd

def add_days_since_first_submission(
    df,
    input_column_name='submission_date',
    output_column_name='days_since_first_submission',
    reference_start_time=None
):
    """
    Adds a column to `df` representing the number of days since
    the first submission (or a provided reference time), based on a given input datetime column.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        input_column_name (str): The name of the column containing datetime values.
        output_column_name (str): The name of the output column to store time difference in days.
        reference_start_time (str or pd.Timestamp, optional): The reference start datetime (in UTC).

    Returns:
        pd.DataFrame: Updated DataFrame with the new column.
    """

    # Convert to datetime if needed
    df[input_column_name] = pd.to_datetime(df[input_column_name], errors='coerce')

    # Ensure datetime is timezone-aware in UTC
    if df[input_column_name].dt.tz is None:
        df[input_column_name] = df[input_column_name].dt.tz_localize('UTC')
    else:
        df[input_column_name] = df[input_column_name].dt.tz_convert('UTC')

    # Ensure reference time is UTC-aware
    if reference_start_time is None:
        reference_start_time = df[input_column_name].min()
    else:
        reference_start_time = pd.to_datetime(reference_start_time, utc=True)

    # Compute difference in days
    df[output_column_name] = (
        (df[input_column_name] - reference_start_time).dt.total_seconds() / (3600 * 24)
    ).astype(np.float32)

    return df




# Data Preparation Function
def prepare_interaction_data(user_profile, recommendations, ratings_matrix, presentation_type_options, audience_type_options):
    """
    Prepare the interaction data for training.
    """
    start_time = pd.to_datetime("2005-12-01 00:00:00")
    # Update your recommendations DataFrame
    recommendations = add_days_since_first_submission(recommendations, reference_start_time=start_time)

    interaction_data = pd.merge(ratings_matrix, user_profile, on='user_id', how='inner')
    recommendations_flat = flatten_recommendation_features(recommendations, presentation_type_options, audience_type_options)
    interaction_data = pd.merge(interaction_data, recommendations_flat, on='article_id', how='inner')

    print(interaction_data.columns)
    ########## TODO should i include popularity
    feature_columns = [
        col for col in interaction_data.columns
        if col not in ['user_id', 'article_id', 'rating', 'location', 'Main issue', 'location_restrictions','popularity', 'submission_date', 'timestamp']
    ]
    X = interaction_data[feature_columns]
    y = interaction_data['rating']

    return interaction_data, X, y, feature_columns

# Utility to One-Hot Encode Features
def flatten_recommendation_features(recommendations, presentation_type_options, audience_type_options):
    """
    One-hot encode `presentation_type` and `audience_type` in recommendations.
    """  
    presentation_columns = {ptype: f"presentation_type_{ptype}" for ptype in presentation_type_options}
    presentation_df = pd.DataFrame(0, index=recommendations.index, columns=presentation_columns.values())
    for i, ptypes in enumerate(recommendations['presentation_type']):
        for ptype in ptypes:
            if ptype in presentation_columns:
                presentation_df.loc[i, presentation_columns[ptype]] = 1

    audience_columns = {atype: f"audience_type_{atype}" for atype in audience_type_options}
    audience_df = pd.DataFrame(0, index=recommendations.index, columns=audience_columns.values())
    for i, atypes in enumerate(recommendations['audience_type']):
        for atype in atypes:
            if atype in audience_columns:
                audience_df.loc[i, audience_columns[atype]] = 1
    df = pd.concat(
        [recommendations.drop(columns=['presentation_type', 'audience_type']), presentation_df, audience_df],
        axis=1
    )
 
    return df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

def prepare_recommendation_system_input_data(
    recommendations_for_user_id,
    num_users,
    num_articles,
    ratings_matrix=None
):
    # Step 1: Generate articles dataset
    recommendations = create_dataset(num_articles)

    # Step 2: Generate users
    user_profiles = generate_user_profiles(num_users, types, locations, preferences_options)

    # Step 3: Generate presentations
    presentations_df = generate_presentations_for_users(user_profiles, presentation_problems)

    # Step 4: Generate rating matrix
    ratings_matrix = generate_ratings_matrix(
        user_profiles, recommendations, presentations_df, ratings_matrix
    )
    start_time = pd.to_datetime("2005-12-01 00:00:00")
    # Update your recommendations DataFrame
    ratings_matrix = add_days_since_first_submission(ratings_matrix, input_column_name='timestamp',
    output_column_name='days_since_last_rating', reference_start_time=start_time)
    print(f"Rating matrix length: {len(ratings_matrix)}")

    # Step 5: Preprocess the user profile
    user_profiles_flat = preprocess_user_profile(
        user_profiles, preferences_options, types, presentation_type_options
    )

    # Step 6: Prepare interaction data
    interaction_data, X, y, feature_columns = prepare_interaction_data(
        user_profiles_flat, recommendations, ratings_matrix,
        presentation_type_options, audience_type_options
    )

    return {
        "user_profiles": user_profiles,
        "user_profiles_flat": user_profiles_flat,
        "recommendations": recommendations,
        "presentations_df": presentations_df,
        "ratings_matrix": ratings_matrix,
        "interaction_data": interaction_data,
        "X": X,
        "y": y,
        "feature_columns": feature_columns
    }
```
