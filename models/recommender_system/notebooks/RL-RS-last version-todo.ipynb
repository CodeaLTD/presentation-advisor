```python
import random
import pandas as pd
from datetime import datetime, timedelta

def generate_presentations_for_users(num_users, problems, max_issues_per_presentation=6):
    """
    Generate presentations for a given number of users with random issues marked.

    Parameters:
        num_users (int): The number of users to generate presentations for.
        problems (list): List of possible problems to mark in presentations.
        max_issues_per_presentation (int): Maximum number of issues per presentation (default: 6).

    Returns:
        pd.DataFrame: A DataFrame containing all generated presentations.
    """
    def generate_random_date():
        """
        Generate a random date within the last two years.

        Returns:
            str: A string representing the random date in 'YYYY-MM-DD' format.
        """
        start_date = pd.to_datetime("2005-12-01 00:00:00")
        end_date = pd.to_datetime("now")
        # Generate a random timestamp within the range
        timestamp = np.random.choice(pd.date_range(start=start_date, end=end_date, freq="H"))
        return timestamp

    def generate_presentation(user_id, presentation_count):
        """
        Generate a random presentation with issues marked for a specific user.

        Parameters:
            user_id (int): The ID of the user.
            presentation_count (int): Counter to ensure unique presentation IDs.

        Returns:
            dict: A dictionary representing a presentation with marked issues.
        """
        # Initialize all problems with 0
        presentation = {problem: 0 for problem in problems}
        # Select a random number of problems to mark as 1
        num_problems = random.randint(1, max_issues_per_presentation)
        selected_problems = random.sample(problems, num_problems)
        for problem in selected_problems:
            presentation[problem] = 1
        presentation['user_id'] = user_id
        presentation['presentation_id'] = f"U{user_id}_P{presentation_count}"
        presentation['submission_date'] = generate_random_date()  # Add random submission date
        return presentation

    presentations = []
    presentation_count = 1  # Initialize a counter for unique presentation IDs
    for user_id in user_profiles['user_id']:
        # Each user gets at least one presentation
        presentations.append(generate_presentation(user_id, presentation_count))
        presentation_count += 1
        # Some users randomly get an additional presentation
        if random.choice([True, False]):
            presentations.append(generate_presentation(user_id, presentation_count))
            presentation_count += 1

    return pd.DataFrame(presentations)

```

```python
import numpy as np
import pandas as pd
import random

def generate_article_ratings_with_preferences_and_popularity(
    user_profiles, recommendations, presentations_df, existing_ratings=None, min_rating=1, max_rating=5
):
    """
    Generate article ratings based on user-linked presentations, issues, audience type,
    preferred presentation type, and article popularity. Each rating includes a timestamp.
    
    Parameters:
        user_profiles (pd.DataFrame): DataFrame containing user profiles.
        recommendations (pd.DataFrame): DataFrame containing article recommendations.
        presentations_df (pd.DataFrame): DataFrame linking users to presentations with issues and submission dates.
        existing_ratings (pd.DataFrame, optional): DataFrame containing existing ratings for articles. Defaults to None.
        min_rating (int): Minimum possible rating value.
        max_rating (int): Maximum possible rating value.
    
    Returns:
        pd.DataFrame: A DataFrame containing user_id, article_id, rating, and timestamp.
    """
    ratings = []
    
    # Convert submission_date to datetime for prioritization
    presentations_df["submission_date"] = pd.to_datetime(presentations_df["submission_date"])
    
    # Calculate average article ratings if existing ratings are provided
    if existing_ratings is not None:
        article_avg_ratings = existing_ratings.groupby("article_id")["rating"].mean()
    else:
        article_avg_ratings = pd.Series(dtype=float)
    
    start_date = pd.to_datetime("2005-12-01 00:00:00")
    end_date = pd.to_datetime("now")
    
    for _, user in user_profiles.iterrows():
        user_id = user["user_id"]
        user_preferences = user["preferences"]
        user_type = user["type"]
        preferred_audience = audience_mapping.get(user_type, "general")
        preferred_presentation_type = user["preferred_presentation_type"]

        # Filter presentations linked to this user
        user_presentations = presentations_df[presentations_df["user_id"] == user_id]
        
        # Weight presentations based on recency
        user_presentations["recency_weight"] = (
            1 / (1 + (pd.Timestamp.now() - user_presentations["submission_date"]).dt.days)
        )
        
        # Aggregate issues for this user's presentations, weighted by recency
        weighted_issues = (
            user_presentations.iloc[:, :-3]
            .multiply(user_presentations["recency_weight"], axis=0)
            .sum()
        )
        
        for _, rec in recommendations.iterrows():
            article_id = rec["article_id"]
            presentation_type = rec["presentation_type"]
            main_issue = rec["Main issue"]
            audience_type = rec["audience_type"]
            
            # Base rating
            rating = (max_rating + min_rating) // 2  # Start with midpoint rating
            
            # Rule 1: Match user preferences with weighted issues
            matching_issues = set(user_preferences) & set(weighted_issues.index[weighted_issues > 0])
            if matching_issues:
                rating += len(matching_issues)  # Increase rating for each matching issue
            
            # Rule 2: Match preferred presentation type
            if preferred_presentation_type == presentation_type:
                rating += 1  # Boost rating for matching presentation type
            else:
                rating -= 0.5  # Penalize for non-matching presentation type
            
            # Rule 3: Match audience type (skip if preferred audience type is "general")
            if preferred_audience != "general":
                if audience_type == preferred_audience:
                    rating += 0.5  # Boost rating for matching audience type
                else:
                    rating -= 0.5  # Penalize for non-matching audience type
            
            # Rule 4: Incorporate main issue
            if main_issue in user_preferences:
                rating += 0.5
            
            # Rule 5: Adjust based on article popularity (average rating)
            if article_id in article_avg_ratings:
                avg_rating = article_avg_ratings[article_id]
                if avg_rating > (max_rating + min_rating) / 2:
                    rating += 0.5  # Boost for higher average ratings
                else:
                    rating -= 0.5  # Penalize for lower average ratings
            
            # Ensure rating is within bounds
            rating = max(min_rating, min(rating, max_rating))
            
            # Generate a random timestamp within the range
            timestamp = np.random.choice(pd.date_range(start=start_date, end=end_date, freq="H"))
            
            # Append the rating
            ratings.append({"user_id": user_id, "article_id": article_id, "rating": rating, "timestamp": timestamp})
    
    return pd.DataFrame(ratings)

def generate_ratings_matrix(user_profiles_df, recommendations, presentations_df, existing_ratings=None):
    """
    Generate the ratings matrix with timestamps.
    """
    return generate_article_ratings_with_preferences_and_popularity(
        user_profiles_df, recommendations, presentations_df, existing_ratings
    )
```

```python
def preprocess_user_profile(user_profile, preferences_options, types, presentation_type_options):
    """
    Preprocess user profile by flattening preferences and binary encoding categorical features.

    Parameters:
        user_profile (pd.DataFrame): DataFrame containing user-specific preferences and attributes.
        preferences_options (list): List of all possible preferences.
        types (list): List of possible user types.
        presentation_type_options (list): List of possible presentation types.

    Returns:
        pd.DataFrame: Updated user_profile with flattened preferences and binary-encoded features.
    """
    # Step 1: Flatten preferences
    preference_columns = {pref: f"user_preference_{pref}" for pref in preferences_options}
    preferences_df = pd.DataFrame(0, index=user_profile.index, columns=preference_columns.values())
    for i, prefs in enumerate(user_profile['preferences']):
        for pref in prefs:
            if pref in preference_columns:
                preferences_df.loc[i, preference_columns[pref]] = 1

    # Step 2: Binary encode 'type'
    type_columns = {t: f"user_type_{t}" for t in types}
    type_df = pd.DataFrame(0, index=user_profile.index, columns=type_columns.values())
    for i, user_type in enumerate(user_profile['type']):
        if user_type in type_columns:
            type_df.loc[i, type_columns[user_type]] = 1

    # Binary encode 'preferred_presentation_type'
    presentation_columns = {pt: f"presentation_type_{pt}" for pt in presentation_type_options}
    presentation_df = pd.DataFrame(0, index=user_profile.index, columns=presentation_columns.values())
    for i, presentation_type in enumerate(user_profile['preferred_presentation_type']):
        if presentation_type in presentation_columns:
            presentation_df.loc[i, presentation_columns[presentation_type]] = 1

    # Concatenate all features with the original DataFrame
    user_profile_encoded = pd.concat([
        user_profile.drop(columns=['preferences', 'type', 'preferred_presentation_type']),
        preferences_df, type_df, presentation_df
    ], axis=1)

    return user_profile_encoded
```

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

def calculate_problem_weights(df, date_column, problem_columns, days_window=90, decay_factor=0.01, apply_filter=True):
    """
    Изчислява претеглена вероятност за всеки проблем на база на презентации от последните days_window дни.
    Важно: Колоната, зададена с date_column, съдържа време в часове от първата подадена презентация.
    :param df: DataFrame с колоните [date_column] и [problem_columns]
    :param date_column: Име на колоната, която съдържа времето (в часове) от първата submission.
    :param problem_columns: Списък с колоните за проблемите (0 или 1)
    :param days_window: Колко дни назад да се вземат презентации (по подразбиране 90 дни)
    :param decay_factor: Фактор на затихване (по подразбиране 0.01)
    :param apply_filter: Ако True, филтрира по времевия прозорец. Ако False, използва целия DataFrame.
    :return: df_filtered (или оригиналния df ако не се филтрира) и NumPy масив с претеглените стойности за всеки проблем.
    """
    
    # Ако DataFrame има само един ред, връщаме директно неговите стойности.
    if len(df) <= 1:
        arr = np.array([df[col].iloc[0] if not df.empty else 0 for col in problem_columns])
        return df, arr

    # Ако apply_filter е True, филтрираме редовете според времевия прозорец.
    # Изчисляваме прага в часове (дни_window * 24)
    if apply_filter:
        threshold_hours = days_window * 24
        df_filtered = df[df[date_column] <= threshold_hours]
    else:
        df_filtered = df.copy()

    # Ако след филтрирането няма редове, връщаме масив със 0 стойности.
    if df_filtered.empty:
        print("No data after filtering. Returning zeros.")
        return df, np.array([0 for _ in problem_columns])
    
    # Изчисляване на тежестите спрямо възрастта на презентациите.
    # Тъй като колоната е в часове, преобразуваме часовете в дни (делим на 24)
    # за да запазим decay_factor, дефиниран за дни.
    df_filtered["weight"] = np.exp(-decay_factor * (df_filtered[date_column] / 24))
    
    # Изчисляване на претегленото средно за всеки проблем
    weighted_problems = np.array([
        (df_filtered[col] * df_filtered["weight"]).sum() / df_filtered["weight"].sum() 
        for col in problem_columns
    ])
    
    return df_filtered, weighted_problems

padded_problem_sequences = []
user_data = {}

df_user_full = data1[user_features].copy()
for user_id_to_fetch in X_user_id:
    if user_id_to_fetch in user_data:
        # If user data is already calculated, use it directly
        weighted_problems = user_data[user_id_to_fetch]
    else:
        # Filter the DataFrame for the current user
        df_user = problem_sequences_df[problem_sequences_df['user_id'] == user_id_to_fetch]
        # Calculate weighted problems
        _, weighted_problems = calculate_problem_weights(df_user, 
                                                          "Time_since_first_submission_hours", 
                                                          problem_features, 
                                                          days_window=180, 
                                                          decay_factor=0.01, 
                                                          apply_filter=False)
        
        # Store the result in user_data
        user_data[user_id_to_fetch] = weighted_problems
        
    
    # Append to padded_problem_sequences
    padded_problem_sequences.append(weighted_problems)
    
padded_problem_sequences = np.array(padded_problem_sequences)
print(padded_problem_sequences.shape)
# If it's 1D  reshape to 2D Keras typically expects a 2D tensor for each input, where:The first dimension (axis 0) represents the number of samples in the batch.The second dimension (axis 1) represents the features for each sample.
#if len(padded_problem_sequences.shape) == 1:
#    padded_problem_sequences = padded_problem_sequences.reshape(-1, 1)

df_new = pd.DataFrame(padded_problem_sequences, columns=problem_features)
df_user_full = pd.concat([df_user_full, df_new], axis=1)
```

```python
import gym
import numpy as np
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from collections import deque

# Custom Gym Environment
class PresentationRecommenderEnv(gym.Env):
    def __init__(self, df, user_data, item_data):
        super(PresentationRecommenderEnv, self).__init__()
        
        self.df = df
        self.user_features = user_data.values.astype(np.float32)
        self.item_features = item_data.values.astype(np.float32)
        self.ratings = df[['user_id', 'article_id', 'rating']].values  # Convert to NumPy
        
        self.articles = np.unique(self.ratings[:, 1])  # Unique article IDs
        self.article_to_idx = {article: idx for idx, article in enumerate(self.articles)}
        self.n_articles = len(self.articles)
        
        self.action_space = gym.spaces.Discrete(self.n_articles)
        self.observation_space = gym.spaces.Box(low=0, high=1, shape=(self.user_features.shape[1] + self.item_features.shape[1],), dtype=np.float32)

        self.current_index = 0
    
    def reset(self):
        self.current_index = 0
        return np.hstack([self.user_features[self.current_index], self.item_features[self.current_index]])

    def step(self, action):
        user_id = self.ratings[self.current_index, 0]
        article_id = self.articles[action]

        # Vectorized lookup for rating instead of using .iloc[]
        mask = (self.ratings[:, 0] == user_id) & (self.ratings[:, 1] == article_id)
        reward = self.ratings[mask, 2].sum() if np.any(mask) else 0
        
        self.current_index += 1
        done = self.current_index >= len(self.df)
        next_state = (
            np.hstack([self.user_features[self.current_index], self.item_features[self.current_index]])
            if not done else np.zeros(self.user_features.shape[1] + self.item_features.shape[1])
        )

        return next_state, reward, done, {}

# Optimized DQN Model
class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        
        self.model = self._build_model()
        self.target_model = self._build_model()  # Target network
        self.update_target_model()

        # 🔥 FIX: Define optimizer
        self.optimizer = keras.optimizers.Adam(learning_rate=self.learning_rate)

    def _build_model(self):
        model = keras.Sequential([
            layers.Dense(256, activation='relu', input_shape=(self.state_size,)),
            layers.Dense(128, activation='relu'),
            layers.Dense(self.action_size, activation='linear')
        ])
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=self.learning_rate), loss='mse')
        return model

    def update_target_model(self):
        """ Update the target network weights. """
        self.target_model.set_weights(self.model.get_weights())

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return np.random.randint(self.action_size)
        q_values = self.model.predict(np.expand_dims(state, axis=0), verbose=0)
        return np.argmax(q_values[0])

    #Temporal Difference Learning
    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return
        
        minibatch = random.sample(self.memory, batch_size)
        states, targets = [], []

        for state, action, reward, next_state, done in minibatch:
            target = reward
            if not done:
                #Bellman update -estimates the future reward.
                # This also follows the TD update rule, where future rewards (max_a Q(s', a)) contribute to current learning.
                target += self.gamma * np.amax(self.target_model.predict(np.expand_dims(next_state, axis=0), verbose=0)[0])
            #Q-Learning is a value-based RL method that updates Q-values based on the Bellman equation. a Q-table stores all (state, action) 
            #values,but we use NN
            target_f = self.model.predict(np.expand_dims(state, axis=0), verbose=0)
            target_f[0][action] = target
            
            states.append(state)
            targets.append(target_f[0])

        # 🔥 FIX: Use train_on_batch instead of gradients manually
        self.model.train_on_batch(np.array(states), np.array(targets))

        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

# Faster Training with Target Network & Vectorized Operations
def train_dqn(df, user_features, item_features, episodes=500, batch_size=32, update_target_every=10):
    env = PresentationRecommenderEnv(df, user_features, item_features)
    agent = DQNAgent(env.observation_space.shape[0], env.action_space.n)
    
    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        done = False
        
        while not done:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            state = next_state
            total_reward += reward

        agent.replay(batch_size)

        # Update target network every few episodes
        if e % update_target_every == 0:
            agent.update_target_model()

        if e % 50 == 0:
            print(f"Episode {e}, Total Reward: {total_reward:.2f}, Epsilon: {agent.epsilon:.4f}")

    return agent

from sklearn.model_selection import train_test_split

# Step 1: Extract user & item features from the full dataset
user_features_full = df_user_full.copy()  # Copy to avoid modifying original
item_features_full = data1[item_features].copy()  # Copy item features

# Step 2: First split: 70% Train, 30% Temporary (Val + Test)
train_data, temp_data = train_test_split(data1, test_size=0.30, random_state=42)

# Step 3: Second split: 50% of temp_data goes to Val, 50% to Test (15% each)
val_data, test_data = train_test_split(temp_data, test_size=0.50, random_state=42)

# Step 4: Match user and item features for each split
train_user_features = user_features_full.loc[train_data.index]
train_item_features = item_features_full.loc[train_data.index]

val_user_features = user_features_full.loc[val_data.index]
val_item_features = item_features_full.loc[val_data.index]

test_user_features = user_features_full.loc[test_data.index]
test_item_features = item_features_full.loc[test_data.index]

# Print data sizes to verify correct splitting
print(f"Train Data: {len(train_data)} rows")
print(f"Validation Data: {len(val_data)} rows")
print(f"Test Data: {len(test_data)} rows")

# Example usage
agent = train_dqn(train_data, train_user_features, train_item_features)
```

```python
import gym
import numpy as np
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from collections import deque
import collections

# Custom Gym Environment
class PresentationRecommenderEnv(gym.Env):
    def __init__(self, df, user_data, item_data):
        super(PresentationRecommenderEnv, self).__init__()
        
        self.df = df
        self.user_features = user_data.values.astype(np.float32)
        self.item_features = item_data.values.astype(np.float32)
        self.ratings = df[['user_id', 'article_id', 'rating']].values  # Convert to NumPy
        
        self.users = np.unique(self.ratings[:, 0])  # Unique user IDs
        self.articles = np.unique(self.ratings[:, 1])  # Unique article IDs
        self.article_to_idx = {article: idx for idx, article in enumerate(self.articles)}
        self.user_to_idx = {user: idx for idx, user in enumerate(self.users)}
        self.n_users = len(self.users)
        self.n_articles = len(self.articles)
        
        self.action_space = gym.spaces.Discrete(self.n_articles)
        self.observation_space = gym.spaces.Box(low=0, high=1, shape=(self.user_features.shape[1] + self.item_features.shape[1],), dtype=np.float32)

        self.current_user = None

    def reset(self):
        self.current_user = np.random.choice(self.users)  # Random user selection
        user_idx = self.user_to_idx[self.current_user]

        return np.hstack([self.user_features[user_idx], np.zeros(self.item_features.shape[1])])

    def step(self, action):
        article_id = self.articles[action]
        mask = (self.ratings[:, 0] == self.current_user) & (self.ratings[:, 1] == article_id)
        reward = self.ratings[mask, 2].mean() if np.any(mask) else -0.1  # Small penalty for wrong choices
        reward = np.clip(reward, -1, 1)  # Normalize rewards

        next_article = np.random.choice(self.articles)
        next_article_idx = self.article_to_idx[next_article]

        next_state = np.hstack([self.user_features[self.user_to_idx[self.current_user]], self.item_features[next_article_idx]])
        done = np.random.rand() < 0.05  # Random episode termination

        return next_state, reward, done, {}

# Double DQN with Prioritized Experience Replay
class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=5000)
        self.gamma = 0.99  # Higher discount factor for long-term rewards
        self.epsilon = 1.0
        self.epsilon_min = 0.05  # Allow minimal exploration
        self.epsilon_decay = 0.99  # Slower decay for better exploration
        self.learning_rate = 0.0005  # Lower learning rate for stability
        self.update_target_every = 500  # Update every 500 steps

        self.model = self._build_model()
        self.target_model = self._build_model()
        self.update_target_model()

    def _build_model(self):
        model = keras.Sequential([
            layers.Dense(256, activation='relu', input_shape=(self.state_size,)),
            layers.Dense(128, activation='relu'),
            layers.Dense(self.action_size, activation='linear')
        ])
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=self.learning_rate), loss='mse')
        return model

    def update_target_model(self):
        self.target_model.set_weights(self.model.get_weights())

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return np.random.randint(self.action_size)
        q_values = self.model.predict(np.expand_dims(state, axis=0), verbose=0)
        return np.argmax(q_values[0])

    def replay(self, batch_size=64):
        if len(self.memory) < batch_size:
            return

        minibatch = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*minibatch)

        states = np.array(states)
        next_states = np.array(next_states)
        targets = self.model.predict(states, verbose=0)

        # Double DQN: Use main model for action selection, target model for Q-value estimation
        best_actions = np.argmax(self.model.predict(next_states, verbose=0), axis=1)
        target_q_values = self.target_model.predict(next_states, verbose=0)
        max_next_q_values = target_q_values[np.arange(batch_size), best_actions]

        for i in range(batch_size):
            targets[i, actions[i]] = rewards[i] if dones[i] else rewards[i] + self.gamma * max_next_q_values[i]

        self.model.train_on_batch(states, targets)

        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

# Training Function with Moving Average Reward Tracking
def train_dqn(df, user_features, item_features, episodes=300, batch_size=64, update_target_every=50):
    env = PresentationRecommenderEnv(df, user_features, item_features)
    agent = DQNAgent(env.observation_space.shape[0], env.action_space.n)
    
    reward_history = collections.deque(maxlen=50)  # Track last 50 episodes
    step_count = 0  # Count total training steps

    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        done = False

        while not done:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            state = next_state
            total_reward += reward
            step_count += 1  # Count steps

            # ✅ Update target network every `update_target_every` steps
            if step_count % update_target_every == 0:
                agent.update_target_model()

        agent.replay(batch_size)  # Train the model after each episode

        # Track the moving average reward
        reward_history.append(total_reward)
        avg_reward = np.mean(reward_history)

        if e % 10 == 0:
            print(f"Episode {e}, Avg Reward: {avg_reward:.2f}, Epsilon: {agent.epsilon:.4f}")

    return agent


# Example Usage
from sklearn.model_selection import train_test_split

# Step 1: Extract user & item features from the full dataset
user_features_full = df_user_full.copy()  # Copy to avoid modifying original
item_features_full = data1[item_features].copy()  # Copy item features

# Step 2: First split: 70% Train, 30% Temporary (Val + Test)
train_data, temp_data = train_test_split(data1, test_size=0.30, random_state=42)

# Step 3: Second split: 50% of temp_data goes to Val, 50% to Test (15% each)
val_data, test_data = train_test_split(temp_data, test_size=0.50, random_state=42)

# Step 4: Match user and item features for each split
train_user_features = user_features_full.loc[train_data.index]
train_item_features = item_features_full.loc[train_data.index]

val_user_features = user_features_full.loc[val_data.index]
val_item_features = item_features_full.loc[val_data.index]

test_user_features = user_features_full.loc[test_data.index]
test_item_features = item_features_full.loc[test_data.index]

# Print data sizes to verify correct splitting
print(f"Train Data: {len(train_data)} rows")
print(f"Validation Data: {len(val_data)} rows")
print(f"Test Data: {len(test_data)} rows")

# Example usage
agent = train_dqn(train_data, train_user_features, train_item_features, 500,64,50)

```

```python
import gym
import numpy as np
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from collections import deque
import collections

# Custom Gym Environment
class PresentationRecommenderEnv(gym.Env):
    def __init__(self, df, user_data, item_data):
        super(PresentationRecommenderEnv, self).__init__()
        
        self.df = df
        self.user_features = user_data.values.astype(np.float32)
        self.item_features = item_data.values.astype(np.float32)
        self.ratings = df[['user_id', 'article_id', 'rating']].values  # Convert to NumPy
        
        self.users = np.unique(self.ratings[:, 0])  # Unique user IDs
        self.articles = np.unique(self.ratings[:, 1])  # Unique article IDs
        self.article_to_idx = {article: idx for idx, article in enumerate(self.articles)}
        self.user_to_idx = {user: idx for idx, user in enumerate(self.users)}
        self.n_users = len(self.users)
        self.n_articles = len(self.articles)
        
        self.action_space = gym.spaces.Discrete(self.n_articles)
        self.observation_space = gym.spaces.Box(low=0, high=1, shape=(self.user_features.shape[1] + self.item_features.shape[1],), dtype=np.float32)

        self.current_user = None

    def reset(self):
        self.current_user = np.random.choice(self.users)  # Random user selection
        user_idx = self.user_to_idx[self.current_user]
        return np.hstack([self.user_features[user_idx], np.zeros(self.item_features.shape[1])])

    def step(self, action):
        article_id = self.articles[action]
        mask = (self.ratings[:, 0] == self.current_user) & (self.ratings[:, 1] == article_id)
        reward = self.ratings[mask, 2].mean() if np.any(mask) else -0.1  # Small penalty for wrong choices
        
        # ✅ Normalize rewards between 0 and 1
        reward = (reward - np.min(self.ratings[:, 2])) / (np.max(self.ratings[:, 2]) - np.min(self.ratings[:, 2]))

        next_article = np.random.choice(self.articles)
        next_article_idx = self.article_to_idx[next_article]

        next_state = np.hstack([self.user_features[self.user_to_idx[self.current_user]], self.item_features[next_article_idx]])
        done = np.random.rand() < 0.05  # Random episode termination

        return next_state, reward, done, {}

# Double DQN with Prioritized Experience Replay
class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=5000)
        self.gamma = 0.99  # Higher discount factor for long-term rewards
        self.epsilon = 1.0
        self.epsilon_min = 0.01  # 🔥 Reduced for better convergence
        self.epsilon_decay = 0.99  # 🔥 Slower decay to avoid early convergence
        self.learning_rate = 0.0005  # Lower for stable training
        self.update_target_every = 200  # 🔥 Faster updates for better alignment

        self.model = self._build_model()
        self.target_model = self._build_model()
        self.update_target_model()

    def _build_model(self):
        model = keras.Sequential([
            layers.Dense(256, activation='relu', input_shape=(self.state_size,)),
            layers.Dense(128, activation='relu'),
            layers.Dense(self.action_size, activation='linear')
        ])
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=self.learning_rate), loss='mse')
        return model

    def update_target_model(self, tau=0.005):
        """ ✅ Soft update of target network weights using Polyak averaging. """
        target_weights = np.array(self.target_model.get_weights(), dtype=object)
        main_weights = np.array(self.model.get_weights(), dtype=object)
        new_weights = tau * main_weights + (1 - tau) * target_weights
        self.target_model.set_weights(new_weights)

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return np.random.randint(self.action_size)
        q_values = self.model.predict(np.expand_dims(state, axis=0), verbose=0)
        return np.argmax(q_values[0])

    def replay(self, batch_size=128):  # 🔥 Increased batch size for stable learning
        if len(self.memory) < batch_size:
            return

        minibatch = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*minibatch)

        states = np.array(states)
        next_states = np.array(next_states)
        targets = self.model.predict(states, verbose=0)

        # ✅ Double DQN: Use main model for action selection, target model for Q-value estimation
        best_actions = np.argmax(self.model.predict(next_states, verbose=0), axis=1)
        target_q_values = self.target_model.predict(next_states, verbose=0)
        max_next_q_values = target_q_values[np.arange(batch_size), best_actions]

        for i in range(batch_size):
            targets[i, actions[i]] = rewards[i] if dones[i] else rewards[i] + self.gamma * max_next_q_values[i]

        self.model.train_on_batch(states, targets)

        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

# Training Function
def train_dqn(df, user_features, item_features, episodes=800):
    env = PresentationRecommenderEnv(df, user_features, item_features)
    agent = DQNAgent(env.observation_space.shape[0], env.action_space.n)
    
    reward_history = collections.deque(maxlen=50)  # Track last 50 episodes
    step_count = 0  # Count total steps

    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        done = False

        while not done:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            state = next_state
            total_reward += reward
            step_count += 1

            if step_count % agent.update_target_every == 0:
                agent.update_target_model()  # ✅ Apply smooth Polyak averaging

        agent.replay(128)  # ✅ Increased batch size

        reward_history.append(total_reward)
        avg_reward = np.mean(reward_history)

        if e % 10 == 0:
            print(f"Episode {e}, Avg Reward: {avg_reward:.2f}, Epsilon: {agent.epsilon:.4f}")

    return agent

# Create the test environment
test_env = PresentationRecommenderEnv(test_data, test_user_features, test_item_features)

# Run the evaluation
def evaluate_agent(agent, env, episodes=50):
    predictions = []
    for _ in range(episodes):
        state = env.reset()
        total_reward = 0
        done = False

        while not done:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            predictions.append((env.current_user, env.articles[action], reward))
            state = next_state
            total_reward += reward

    print(f"Total Reward on Test Set: {total_reward:.2f}")
    return predictions

# Train the agent
agent = train_dqn(train_data, train_user_features, train_item_features)

# Evaluate
test_predictions = evaluate_agent(agent, test_env)

# Convert predictions to a DataFrame
import pandas as pd
test_predictions_df = pd.DataFrame(test_predictions, columns=["user_id", "article_id", "predicted_rating"])
```

```python
from sklearn.metrics import mean_squared_error

def evaluate_mse(agent, df_test, user_features, item_features):
    """
    Compute the MSE of the agent's predictions on a test dataset.

    Args:
    - agent: Trained DQNAgent
    - df_test: Pandas DataFrame (test dataset with user_id, article_id, rating)
    - user_features: List of column names for user features
    - item_features: List of column names for item features

    Returns:
    - MAE, MSE, RMSE score
    """
    y_true = []  # Actual ratings
    y_pred = []  # Predicted Q-values

    for _, row in df_test.iterrows():
        user_f = row[user_features].values  # Extract user feature vector
        item_f = row[item_features].values  # Extract item feature vector

        # Create state input (concatenate user + item features)
        state = np.hstack([user_f, item_f]).reshape(1, -1)

        # Get predicted Q-value (expected rating)
        q_value = agent.model.predict(state, verbose=0)[0][0]

        y_true.append(row['rating'])  # Actual rating
        y_pred.append(q_value)  # Predicted rating (Q-value)

        # Calculate error metrics
        differences = y_pred - y_true
        absolute_differences = np.abs(differences)
        
        mae = np.mean(absolute_differences)
        mse = np.mean(differences ** 2)
        rmse = np.sqrt(mse)
    
        # Print results
        print("Mean Absolute Error (MAE):", mae)
        print("Mean Squared Error (MSE):", mse)
        print("Root Mean Squared Error (RMSE):", rmse)

    return mae, mse, rmse

# Example Usage
mae, mse, rmse = evaluate_mse(agent, test_data, test_user_features, test_item_features)
print(f"Model MSE: {mse:.4f}")
```

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, Concatenate, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

def build_model(user_features, item_features, time_features, padded_problem_sequences):
    """
    Builds and compiles the model.
    
    :param user_features: NumPy array of user features.
    :param item_features: NumPy array of item features.
    :param time_features: NumPy array of temporal features.
    :param padded_problem_sequences: NumPy array of problem sequence features.
    :return: A compiled Keras model.
    """
    # Define input layers
    user_feat_in = Input(shape=(user_features.shape[1],), name="user_feat_input")
    item_feat_in = Input(shape=(item_features.shape[1],), name="item_feat_input")
    time_in = Input(shape=(time_features.shape[1],), name="time_input")
    problem_seq_in = Input(shape=(padded_problem_sequences.shape[1],), name="problem_seq_input")
    
    # Regularization
    regularizer = l2(1e-4)
    
    # User Tower
    user_out = Dense(64, activation="relu", kernel_regularizer=regularizer)(user_feat_in)
    user_out = BatchNormalization()(user_out)
    user_out = Dropout(0.3)(user_out)
    user_emb = Dense(32, activation="relu", kernel_regularizer=regularizer, name="user_embedding")(user_out)
    
    # Item Tower
    item_out = Dense(64, activation="relu", kernel_regularizer=regularizer)(item_feat_in)
    item_out = BatchNormalization()(item_out)
    item_out = Dropout(0.3)(item_out)
    item_emb = Dense(32, activation="relu", kernel_regularizer=regularizer, name="item_embedding")(item_out)
    
    # Temporal Features
    time_out = Dense(8, activation="relu", kernel_regularizer=regularizer)(time_in)
    
    # Problem Sequence Input (processed via a Dense layer)
    problem_out = Dense(32, activation="relu", kernel_regularizer=regularizer)(problem_seq_in)
    
    # Final concatenation of all towers
    merged = Concatenate()([user_emb, item_emb, time_out, problem_out])
    merged = Dense(64, activation="relu", kernel_regularizer=regularizer)(merged)
    merged = Dropout(0.4)(merged)
    output = Dense(1, activation="linear")(merged)
    
    # Build and compile the model
    model = Model(inputs=[user_feat_in, item_feat_in, time_in, problem_seq_in], outputs=output)
    model.compile(optimizer=Adam(learning_rate=0.0005), loss="mse", metrics=["mae"])
    
    return model

def train_model(train_data, val_data, user_features, item_features, time_features, padded_problem_sequences):
    """
    Trains the model using the provided training and validation data.
    
    :param train_data: Tuple/list containing training data in the following order:
                       (train_user_features, train_item_features, train_time_features, train_problem_seq, train_labels)
    :param val_data: Tuple/list containing validation data in the following order:
                     (val_user_features, val_item_features, val_time_features, val_problem_seq, val_labels)
    :param user_features: NumPy array of user features (used for input shape).
    :param item_features: NumPy array of item features (used for input shape).
    :param time_features: NumPy array of temporal features (used for input shape).
    :param padded_problem_sequences: NumPy array of problem sequence features (used for input shape).
    :return: A tuple containing the trained model and the training history.
    """
    # Build the model
    model = build_model(user_features, item_features, time_features, padded_problem_sequences)
    
    # Define callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
    
    # Unpack training and validation data
    (train_user_features, train_item_features, train_time_features,
     train_problem_seq, train_labels) = train_data
    (val_user_features, val_item_features, val_time_features,
     val_problem_seq, val_labels) = val_data
    
    # Train the model
    history = model.fit(
        [train_user_features, train_item_features, train_time_features, train_problem_seq],
        train_labels,
        validation_data=(
            [val_user_features, val_item_features, val_time_features, val_problem_seq],
            val_labels
        ),
        epochs=200,
        batch_size=32,
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    
    return model, history

# Example usage:
# Assume that the following variables are already defined as NumPy arrays (or can be converted to such):
# - train_user_features, train_item_features, train_time_features, train_problem_seq, train_labels
# - val_user_features, val_item_features, val_time_features, val_problem_seq, val_labels
# - user_features, item_features, time_features, padded_problem_sequences (for input shapes)

# Pack your training and validation data into tuples/lists:
train_data = (train_user_features, train_item_features, train_time_features, train_problem_seq, train_labels)
val_data   = (val_user_features, val_item_features, val_time_features, val_problem_seq, val_labels)

# Then call the training function:
model, history = train_model(train_data, val_data, user_features, item_features, time_features, padded_problem_sequences)
```

```python
import numpy as np
import pandas as pd

def sort_by_preferred_presentation(recommendations, user_profile):
    """
    Rearrange recommendations so that those with the user's preferred presentation type appear first.
    
    Parameters:
      - recommendations (pd.DataFrame): DataFrame with a column 'presentation_type'.
      - user_profile (dict): Contains the key 'preferred_presentation_type'.
      
    Returns:
      - pd.DataFrame: Recommendations sorted with preferred type at the top.
    """
    preferred_type = user_profile.get('preferred_presentation_type')
    # Create a helper boolean column.
    recommendations['is_preferred'] = recommendations['presentation_type'] == preferred_type
    sorted_recs = recommendations.sort_values(by='is_preferred', ascending=False).drop(columns='is_preferred')
    return sorted_recs

def limit_recommendations_by_engagement_time(recommendations, engagement_time):
    """
    Limit the number of recommendations based on the user's engagement time.
    
    For example, assume that 5 minutes of engagement allows for 1 recommendation.
    
    Parameters:
      - recommendations (pd.DataFrame): Sorted recommendations.
      - engagement_time (float): User engagement time in minutes.
    
    Returns:
      - pd.DataFrame: The top recommendations limited by engagement.
    """
    max_recs = int(engagement_time / 5)  # 5 minutes = 1 recommendation
    return recommendations.head(max_recs)

def audience_type_decay(row, user_profile):
    """
    Adjust the predicted rating (as the base score) by boosting it if the article's audience type
    matches the user's audience type.
    
    Parameters:
      - row (pd.Series): A row from the recommendations DataFrame.
      - user_profile (dict): Contains the key 'audience_type'.
    
    Returns:
      - float: The adjusted score.
    """
    # Use the predicted rating as the base.
    score = row.get('predicted_rating', 0)
    # Boost the score if audience types match.
    if row.get('audience_type') == user_profile.get('audience_type'):
        score *= 1.1  # Boost factor
    return score

def finalize_recommendations(predicted_articles, user_profile, engagement_time):
    """
    Apply sorting by presentation type, compute an adjusted utility score based on the predicted rating
    and audience type matching, remove duplicate articles, and limit the recommendations based on the user's
    engagement time.
    
    Parameters:
      - predicted_articles (pd.DataFrame): Must include columns:
          'article_id', 'predicted_rating', 'presentation_type', 'audience_type'
      - user_profile (dict): Contains 'preferred_presentation_type' and 'audience_type'.
      - engagement_time (float): The user's engagement time in minutes.
    
    Returns:
      - pd.DataFrame: The final sorted and limited recommendations.
    """
    # Sort by user's preferred presentation type.
    recs_sorted = sort_by_preferred_presentation(predicted_articles.copy(), user_profile)
    
    # Compute adjusted score using audience type decay.
    recs_sorted['utility_score'] = recs_sorted.apply(
        lambda row: audience_type_decay(row, user_profile), axis=1
    )
    
    # Sort recommendations by utility score in descending order.
    recs_sorted = recs_sorted.sort_values('utility_score', ascending=False)
    
    # Remove duplicate articles based on 'article_id'
    recs_sorted = recs_sorted.drop_duplicates(subset='article_id', keep='first')
    
    # Limit the recommendations based on the user's engagement time.
    final_recs = limit_recommendations_by_engagement_time(recs_sorted, engagement_time)
    
    return final_recs

# ------------------------------------------------------------------------------
# Example Usage:
# ------------------------------------------------------------------------------

# Create a dummy predicted_articles DataFrame with the required columns.
dummy_data = {
    'article_id': range(1, 501),
    'predicted_rating': np.random.uniform(1.0, 5.0, 500),
    'presentation_type': np.random.choice(['video', 'text', 'interactive'], 500),
    'audience_type': np.random.choice(['professional', 'student'], 500)
}
predicted_articles = pd.DataFrame(dummy_data)

# Define a sample user profile.
user_profile = {
    'preferred_presentation_type': 'video',
    'audience_type': 'professional'
}

# Define the user's engagement time in minutes.
engagement_time = 30  # e.g., 30 minutes

# Get the final recommendations.
final_recommendations = finalize_recommendations(predicted_articles, user_profile, engagement_time)

print("Final Recommendations (top items):")
print(final_recommendations.head(100))
```
