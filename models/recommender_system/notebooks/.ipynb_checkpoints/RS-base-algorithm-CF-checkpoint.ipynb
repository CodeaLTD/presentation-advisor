```python

import random

# Function to get a random name and summary based on the main issue for the articles
def get_name_and_summary(issue):
    if issue in issue_based_data:
        name = random.choice(issue_based_data[issue]["names"])
        summary = random.choice(issue_based_data[issue]["summaries"])
    else:
        name = "General Presentation Tips"
        summary = "Basic guidelines for an effective presentation."
    return name, summary

# Example of generating 10 entries based on main issues for the articles
def generate_entries(n ,sample_main_issues,issue_based_data):
  data = []
  for _ in range(n):
      main_issue = random.choice(sample_main_issues)
      name, summary = get_name_and_summary(main_issue)
      entry = {
          "Main issue": main_issue,
          "Name": name,
          "Summary": summary
      }
      data.append(entry)
  return data

# Display generated data
#generate_entries(10 ,sample_main_issues,issue_based_data)
```

```python
import random
import pandas as pd

# Function to generate a consistent entry for a specific main issue for the articles
def generate_entry(main_issue, article_id, presentation_type=None, audience_type=None, location_restrictions=None):
    # Basic attributes with default values (0)
    entry = {
        "Boring presentation": 0, "Graphics": 0, "Readability": 0, "Consistency": 0,
        "Images": 0, "Bullets": 0, "Text size": 0, "Text heavy": 0, "Tables": 0, "Agenda": 0,
        "Infographics": 0, "Positioning": 0, "Presentation length": 0,
        "General tips": 0, "Presentation skills": 0, "Powerpoint knowledge": 0,
        "tips and tricks": 0, "Explanation of the problem": 0,
        "presentation_type": presentation_type if presentation_type else random.choice(presentation_type_options),
        "audience_type": audience_type if audience_type else random.choice(audience_type_options),
        "location_restrictions": location_restrictions if location_restrictions else random.choice(location_restrictions_options)
    }

    # Apply the attribute settings for the main issue
    if main_issue in attribute_settings:
        for attr, value in attribute_settings[main_issue].items():
            entry[attr] = value

    # Add randomly generated values for any remaining attributes not explicitly set
    for key in entry:
        if entry[key] == 0:  # Only randomize if not already set
            entry[key] = random.randint(0, 1)

    entry["Main issue"] = main_issue
    entry["article_id"] = article_id
    entry['popularity'] =  round(random.uniform(0, 5), 2)
    return entry

# Example of generating a consistent entry for a specific main issue
#main_issue = random.choice(list(attribute_settings.keys()))
#entry = generate_entry(main_issue)

#entry  # Display the generated entry
```

```python
import pandas as pd

# Function to create a dataset of atricles by specifying the number of articles.
def create_dataset(num_articles):
    dataset = []
    sample_main_issues = list(issue_based_data.keys())
    for article_id in range(1, num_articles + 1):
        main_issue = random.choice(sample_main_issues)
        entry = generate_entry(main_issue, article_id)
        dataset.append(entry)
    return pd.DataFrame(dataset)
```

```python
import random
import pandas as pd

def generate_user_profiles(num_users, types, locations, preferences_options):
    """
    Generate user profiles with specified preferences.

    Parameters:
        num_users (int): Number of user profiles to generate.
        types (list): List of user types.
        locations (list): List of locations.
        preferences_options (list): List of preference options.

    Returns:
        pd.DataFrame: DataFrame containing the generated user profiles.
    """
    user_profiles = []
    for user_id in range(1, num_users + 1):
        user_type = random.choice(types)
        location = random.choice(locations)
        preferences = [
          random.choice(preferences_options) for _ in range(random.randint(1, 4))  # Randomly decide the number of preferences (1 to 4)
        ]
        user_profiles.append({
            'user_id': user_id,
            'type': user_type,
            'location': location,
            'preferences': preferences,
            'preferred_presentation_type': random.choice(['formal', 'creative', 'students', 'kids'])
        })
    return pd.DataFrame(user_profiles)
```

```python
import random
import pandas as pd
from datetime import datetime, timedelta

def generate_presentations_for_users(num_users, problems, max_issues_per_presentation=6):
    """
    Generate presentations for a given number of users with random issues marked.

    Parameters:
        num_users (int): The number of users to generate presentations for.
        problems (list): List of possible problems to mark in presentations.
        max_issues_per_presentation (int): Maximum number of issues per presentation (default: 6).

    Returns:
        pd.DataFrame: A DataFrame containing all generated presentations.
    """
    def generate_random_date():
        """
        Generate a random date within the last two years.

        Returns:
            str: A string representing the random date in 'YYYY-MM-DD' format.
        """
        start_date = pd.to_datetime("2005-12-01 00:00:00")
        end_date = pd.to_datetime("now")
        # Generate a random timestamp within the range
        timestamp = np.random.choice(pd.date_range(start=start_date, end=end_date, freq="H"))
        return timestamp

    def generate_presentation(user_id, presentation_count):
        """
        Generate a random presentation with issues marked for a specific user.

        Parameters:
            user_id (int): The ID of the user.
            presentation_count (int): Counter to ensure unique presentation IDs.

        Returns:
            dict: A dictionary representing a presentation with marked issues.
        """
        # Initialize all problems with 0
        presentation = {problem: 0 for problem in problems}
        # Select a random number of problems to mark as 1
        num_problems = random.randint(1, max_issues_per_presentation)
        selected_problems = random.sample(problems, num_problems)
        for problem in selected_problems:
            presentation[problem] = 1
        presentation['user_id'] = user_id
        presentation['presentation_id'] = f"U{user_id}_P{presentation_count}"
        presentation['submission_date'] = generate_random_date()  # Add random submission date
        return presentation

    presentations = []
    presentation_count = 1  # Initialize a counter for unique presentation IDs
    for user_id in user_profiles['user_id']:
        # Each user gets at least one presentation
        presentations.append(generate_presentation(user_id, presentation_count))
        presentation_count += 1
        # Some users randomly get an additional presentation
        if random.choice([True, False]):
            presentations.append(generate_presentation(user_id, presentation_count))
            presentation_count += 1

    return pd.DataFrame(presentations)

```

```python
import numpy as np
import pandas as pd
import random

def generate_article_ratings_with_preferences_and_popularity(
    user_profiles, recommendations, presentations_df, existing_ratings=None, min_rating=1, max_rating=5
):
    """
    Generate article ratings based on user-linked presentations, issues, audience type,
    preferred presentation type, and article popularity. Each rating includes a timestamp.
    
    Parameters:
        user_profiles (pd.DataFrame): DataFrame containing user profiles.
        recommendations (pd.DataFrame): DataFrame containing article recommendations.
        presentations_df (pd.DataFrame): DataFrame linking users to presentations with issues and submission dates.
        existing_ratings (pd.DataFrame, optional): DataFrame containing existing ratings for articles. Defaults to None.
        min_rating (int): Minimum possible rating value.
        max_rating (int): Maximum possible rating value.
    
    Returns:
        pd.DataFrame: A DataFrame containing user_id, article_id, rating, and timestamp.
    """
    ratings = []
    
    # Convert submission_date to datetime for prioritization
    presentations_df["submission_date"] = pd.to_datetime(presentations_df["submission_date"])
    
    # Calculate average article ratings if existing ratings are provided
    if existing_ratings is not None:
        article_avg_ratings = existing_ratings.groupby("article_id")["rating"].mean()
    else:
        article_avg_ratings = pd.Series(dtype=float)
    
    start_date = pd.to_datetime("2005-12-01 00:00:00")
    end_date = pd.to_datetime("now")
    
    for _, user in user_profiles.iterrows():
        user_id = user["user_id"]
        user_preferences = user["preferences"]
        user_type = user["type"]
        preferred_audience = audience_mapping.get(user_type, "general")
        preferred_presentation_type = user["preferred_presentation_type"]

        # Filter presentations linked to this user
        user_presentations = presentations_df[presentations_df["user_id"] == user_id]
        
        # Weight presentations based on recency
        user_presentations["recency_weight"] = (
            1 / (1 + (pd.Timestamp.now() - user_presentations["submission_date"]).dt.days)
        )
        
        # Aggregate issues for this user's presentations, weighted by recency
        weighted_issues = (
            user_presentations.iloc[:, :-3]
            .multiply(user_presentations["recency_weight"], axis=0)
            .sum()
        )
        
        for _, rec in recommendations.iterrows():
            article_id = rec["article_id"]
            presentation_type = rec["presentation_type"]
            main_issue = rec["Main issue"]
            audience_type = rec["audience_type"]
            
            # Base rating
            rating = (max_rating + min_rating) // 2  # Start with midpoint rating
            
            # Rule 1: Match user preferences with weighted issues
            matching_issues = set(user_preferences) & set(weighted_issues.index[weighted_issues > 0])
            if matching_issues:
                rating += len(matching_issues)  # Increase rating for each matching issue
            
            # Rule 2: Match preferred presentation type
            if preferred_presentation_type == presentation_type:
                rating += 1  # Boost rating for matching presentation type
            else:
                rating -= 0.5  # Penalize for non-matching presentation type
            
            # Rule 3: Match audience type (skip if preferred audience type is "general")
            if preferred_audience != "general":
                if audience_type == preferred_audience:
                    rating += 0.5  # Boost rating for matching audience type
                else:
                    rating -= 0.5  # Penalize for non-matching audience type
            
            # Rule 4: Incorporate main issue
            if main_issue in user_preferences:
                rating += 0.5
            
            # Rule 5: Adjust based on article popularity (average rating)
            if article_id in article_avg_ratings:
                avg_rating = article_avg_ratings[article_id]
                if avg_rating > (max_rating + min_rating) / 2:
                    rating += 0.5  # Boost for higher average ratings
                else:
                    rating -= 0.5  # Penalize for lower average ratings
            
            # Ensure rating is within bounds
            rating = max(min_rating, min(rating, max_rating))
            
            # Generate a random timestamp within the range
            timestamp = np.random.choice(pd.date_range(start=start_date, end=end_date, freq="H"))
            
            # Append the rating
            ratings.append({"user_id": user_id, "article_id": article_id, "rating": rating, "timestamp": timestamp})
    
    return pd.DataFrame(ratings)

def generate_ratings_matrix(user_profiles_df, recommendations, presentations_df, existing_ratings=None):
    """
    Generate the ratings matrix with timestamps.
    """
    return generate_article_ratings_with_preferences_and_popularity(
        user_profiles_df, recommendations, presentations_df, existing_ratings
    )
```

```python
def preprocess_user_profile(user_profile, preferences_options, types, presentation_type_options):
    """
    Preprocess user profile by flattening preferences and binary encoding categorical features.

    Parameters:
        user_profile (pd.DataFrame): DataFrame containing user-specific preferences and attributes.
        preferences_options (list): List of all possible preferences.
        types (list): List of possible user types.
        presentation_type_options (list): List of possible presentation types.

    Returns:
        pd.DataFrame: Updated user_profile with flattened preferences and binary-encoded features.
    """
    # Step 1: Flatten preferences
    preference_columns = {pref: f"user_preference_{pref}" for pref in preferences_options}
    preferences_df = pd.DataFrame(0, index=user_profile.index, columns=preference_columns.values())
    for i, prefs in enumerate(user_profile['preferences']):
        for pref in prefs:
            if pref in preference_columns:
                preferences_df.loc[i, preference_columns[pref]] = 1

    # Step 2: Binary encode 'type'
    type_columns = {t: f"user_type_{t}" for t in types}
    type_df = pd.DataFrame(0, index=user_profile.index, columns=type_columns.values())
    for i, user_type in enumerate(user_profile['type']):
        if user_type in type_columns:
            type_df.loc[i, type_columns[user_type]] = 1

    # Binary encode 'preferred_presentation_type'
    presentation_columns = {pt: f"presentation_type_{pt}" for pt in presentation_type_options}
    presentation_df = pd.DataFrame(0, index=user_profile.index, columns=presentation_columns.values())
    for i, presentation_type in enumerate(user_profile['preferred_presentation_type']):
        if presentation_type in presentation_columns:
            presentation_df.loc[i, presentation_columns[presentation_type]] = 1

    # Concatenate all features with the original DataFrame
    user_profile_encoded = pd.concat([
        user_profile.drop(columns=['preferences', 'type', 'preferred_presentation_type']),
        preferences_df, type_df, presentation_df
    ], axis=1)

    return user_profile_encoded
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam



# Data Preparation Function
def prepare_interaction_data(user_profile, recommendations, ratings_matrix, presentation_type_options, audience_type_options):
    """
    Prepare the interaction data for training.
    """
    interaction_data = pd.merge(ratings_matrix, user_profile, on='user_id', how='inner')
    recommendations_flat = flatten_recommendation_features(recommendations, presentation_type_options, audience_type_options)
    interaction_data = pd.merge(interaction_data, recommendations_flat, on='article_id', how='inner')

    print(interaction_data.columns)
    feature_columns = [
        col for col in interaction_data.columns
        if col not in ['user_id', 'article_id', 'rating', 'location', 'Main issue', 'location_restrictions','popularity']
    ]
    X = interaction_data[feature_columns]
    y = interaction_data['rating']

    return interaction_data, X, y, feature_columns

# Utility to One-Hot Encode Features
def flatten_recommendation_features(recommendations, presentation_type_options, audience_type_options):
    """
    One-hot encode `presentation_type` and `audience_type` in recommendations.
    """  
    presentation_columns = {ptype: f"presentation_type_{ptype}" for ptype in presentation_type_options}
    presentation_df = pd.DataFrame(0, index=recommendations.index, columns=presentation_columns.values())
    for i, ptypes in enumerate(recommendations['presentation_type']):
        for ptype in ptypes:
            if ptype in presentation_columns:
                presentation_df.loc[i, presentation_columns[ptype]] = 1

    audience_columns = {atype: f"audience_type_{atype}" for atype in audience_type_options}
    audience_df = pd.DataFrame(0, index=recommendations.index, columns=audience_columns.values())
    for i, atypes in enumerate(recommendations['audience_type']):
        for atype in atypes:
            if atype in audience_columns:
                audience_df.loc[i, audience_columns[atype]] = 1
    df = pd.concat(
        [recommendations.drop(columns=['presentation_type', 'audience_type']), presentation_df, audience_df],
        axis=1
    )
 
    return df
```

```python
import numpy as np
import pandas as pd

def extract_time_features(dates, start_time):
    df_temp = pd.DataFrame({'Datetime': pd.to_datetime(dates)})
    df_temp = df_temp.sort_values(by='Datetime').reset_index(drop=True)
    df_temp['Hour'] = df_temp['Datetime'].dt.hour
    df_temp['Day_of_Week'] = df_temp['Datetime'].dt.dayofweek
    df_temp['Hour_sin'] = np.sin(2 * np.pi * df_temp['Hour'] / 24)
    df_temp['Hour_cos'] = np.cos(2 * np.pi * df_temp['Hour'] / 24)
    df_temp['Day_sin'] = np.sin(2 * np.pi * df_temp['Day_of_Week'] / 7)
    df_temp['Day_cos'] = np.cos(2 * np.pi * df_temp['Day_of_Week'] / 7)
    #start_time = df_temp['Datetime'].min()
    df_temp['Time_since_first_submission_hours'] = (df_temp['Datetime'] - start_time).dt.total_seconds() / 3600
    return df_temp[['Hour_sin', 'Hour_cos', 'Day_sin', 'Day_cos', 'Time_since_first_submission_hours']]
    
start_time = pd.to_datetime("2005-12-01 00:00:00")
time_features=extract_time_features(data1['timestamp'].copy(), start_time)

X_user_id, X_item_id = data1['user_id'].to_numpy(), data1['article_id'].to_numpy()

# Extracting the problem-related features
presentation_df_temp = presentations_df.sort_values(by='user_id').reset_index(drop=True)  
presentation_submission_date_df = extract_time_features(presentation_df_temp['submission_date'], start_time)
problem_sequences_df = pd.concat([presentations_df['user_id'],presentations_df[problem_features], presentation_submission_date_df], axis=1)

#problem_sequences = presentation_df[problem_features].to_numpy()
y=data1['rating']
```

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

def calculate_problem_weights(df, date_column, problem_columns, days_window=90, decay_factor=0.01, apply_filter=True):
    """
    Изчислява претеглена вероятност за всеки проблем на база на презентации от последните days_window дни.
    Важно: Колоната, зададена с date_column, съдържа време в часове от първата подадена презентация.
    :param df: DataFrame с колоните [date_column] и [problem_columns]
    :param date_column: Име на колоната, която съдържа времето (в часове) от първата submission.
    :param problem_columns: Списък с колоните за проблемите (0 или 1)
    :param days_window: Колко дни назад да се вземат презентации (по подразбиране 90 дни)
    :param decay_factor: Фактор на затихване (по подразбиране 0.01)
    :param apply_filter: Ако True, филтрира по времевия прозорец. Ако False, използва целия DataFrame.
    :return: df_filtered (или оригиналния df ако не се филтрира) и NumPy масив с претеглените стойности за всеки проблем.
    """
    
    # Ако DataFrame има само един ред, връщаме директно неговите стойности.
    if len(df) <= 1:
        arr = np.array([df[col].iloc[0] if not df.empty else 0 for col in problem_columns])
        return df, arr

    # Ако apply_filter е True, филтрираме редовете според времевия прозорец.
    # Изчисляваме прага в часове (дни_window * 24)
    if apply_filter:
        threshold_hours = days_window * 24
        df_filtered = df[df[date_column] <= threshold_hours]
    else:
        df_filtered = df.copy()

    # Ако след филтрирането няма редове, връщаме масив със 0 стойности.
    if df_filtered.empty:
        print("No data after filtering. Returning zeros.")
        return df, np.array([0 for _ in problem_columns])
    
    # Изчисляване на тежестите спрямо възрастта на презентациите.
    # Тъй като колоната е в часове, преобразуваме часовете в дни (делим на 24)
    # за да запазим decay_factor, дефиниран за дни.
    df_filtered["weight"] = np.exp(-decay_factor * (df_filtered[date_column] / 24))
    
    # Изчисляване на претегленото средно за всеки проблем
    weighted_problems = np.array([
        (df_filtered[col] * df_filtered["weight"]).sum() / df_filtered["weight"].sum() 
        for col in problem_columns
    ])
    
    return df_filtered, weighted_problems

padded_problem_sequences = []
user_data = {}

df_user_full = data1[user_features].copy()
for user_id_to_fetch in X_user_id:
    if user_id_to_fetch in user_data:
        # If user data is already calculated, use it directly
        weighted_problems = user_data[user_id_to_fetch]
    else:
        # Filter the DataFrame for the current user
        df_user = problem_sequences_df[problem_sequences_df['user_id'] == user_id_to_fetch]
        # Calculate weighted problems
        _, weighted_problems = calculate_problem_weights(df_user, 
                                                          "Time_since_first_submission_hours", 
                                                          problem_features, 
                                                          days_window=180, 
                                                          decay_factor=0.01, 
                                                          apply_filter=False)
        
        # Store the result in user_data
        user_data[user_id_to_fetch] = weighted_problems
        
    
    # Append to padded_problem_sequences
    padded_problem_sequences.append(weighted_problems)
    
padded_problem_sequences = np.array(padded_problem_sequences)
print(padded_problem_sequences.shape)
# If it's 1D  reshape to 2D Keras typically expects a 2D tensor for each input, where:The first dimension (axis 0) represents the number of samples in the batch.The second dimension (axis 1) represents the features for each sample.
#if len(padded_problem_sequences.shape) == 1:
#    padded_problem_sequences = padded_problem_sequences.reshape(-1, 1)

df_new = pd.DataFrame(padded_problem_sequences, columns=problem_features)
df_user_full = pd.concat([df_user_full, df_new], axis=1)
```

```python
import numpy as np

def evaluate_predictions(y_pred, y_test):
    """
    Evaluates predictions using Mean Absolute Error (MAE), 
    Mean Squared Error (MSE), and Root Mean Squared Error (RMSE).

    Parameters:
    - y_pred: Predicted values (numpy array or list).
    - y_test: True target values (numpy array or list).

    Returns:
    - mae: Mean Absolute Error
    - mse: Mean Squared Error
    - rmse: Root Mean Squared Error
    """
    # Ensure arrays are flattened for consistency
    y_pred = np.array(y_pred).flatten()
    y_test = np.array(y_test).flatten()

    # Calculate error metrics
    differences = y_pred - y_test
    absolute_differences = np.abs(differences)
    
    mae = np.mean(absolute_differences)
    mse = np.mean(differences ** 2)
    rmse = np.sqrt(mse)

    # Print results
    print("Mean Absolute Error (MAE):", mae)
    print("Mean Squared Error (MSE):", mse)
    print("Root Mean Squared Error (RMSE):", rmse)

    return mae, mse, rmse
    
evaluate_predictions(final_score['score'], final_score['rating'])
```

```python
import pandas as pd

def extract_user_profile(temp_df: pd.DataFrame):
    # Define columns to extract
    columns_to_extract = [
        'user_id', 'article_id', 'pred',
        'presentation_type_formal_x', 'presentation_type_creative_x', 'presentation_type_students_x',
        'presentation_type_technical_x', 'presentation_type_branding_x',
        'audience_type_academic', 'audience_type_business',
        'audience_type_general', 'audience_type_specialized',
        'time_since'
    ]
    
    # Extract relevant data
    extracted_df = temp_df[columns_to_extract]
    
    # Extract a sample user profile (first user in dataframe)
    first_user = extracted_df.iloc[0]
    
    # Identify the preferred presentation type
    preferred_presentation_type = next(
        (col.replace('presentation_type_', '').replace('_x', '') 
         for col in columns_to_extract[3:8] if first_user[col] == 1),
        None
    )
    
    # Identify the preferred audience type
    preferred_audience_type = next(
        (col.replace('audience_type_', '') 
         for col in columns_to_extract[8:12] if first_user[col] == 1),
        None
    )
    
    # Construct user profile dictionary
    user_profile = {
        'user_id': int(first_user['user_id']),
        'preferred_presentation_type': preferred_presentation_type,
        'audience_type': preferred_audience_type
    }
    
    return extracted_df, user_profile

# Example usage:
extracted_df, user_profile = extract_user_profile(temp_df)
# print(extracted_df.head())
# print(user_profile)

def extract_columns(df: pd.DataFrame) -> pd.DataFrame:
    """
    Extracts specific columns from the given DataFrame.
    
    Parameters:
        df (pd.DataFrame): The input DataFrame.
    
    Returns:
        pd.DataFrame: A new DataFrame containing only the specified columns.
    """
    columns_to_extract = [
        'user_id',
        'article_id',
        'rating',
        'pred',
        'Boring presentation',
         'Graphics',
         'Readability',
         'Consistency',
         'Images',
         'Bullets',
         'Text size',
         'Text heavy',
         'Tables',
         'Agenda',
         'Infographics',
         'Positioning',
         'Presentation length',
         'General tips',
         'Presentation skills',
         'Powerpoint knowledge',
         'tips and tricks',
         'Explanation of the problem',
         'presentation_type_formal_y',
         'presentation_type_creative_y',
         'presentation_type_students_y',
         'presentation_type_technical_y',
         'presentation_type_branding_y',
         'audience_type_academic',
         'audience_type_business',
         'audience_type_kids',
         'audience_type_general',
         'audience_type_creative',
         'audience_type_specialized',
        'time_since'
    ]

    return df[columns_to_extract]

extracted_df = extract_columns(temp_df)
```

```python
import numpy as np
import pandas as pd

#add Long-Tail Boosting: introduces a weighting mechanism to prioritize less popular items, encouraging their recommendation.

def sort_by_preferred_presentation(recommendations, user_profile):
    """
    Sort recommendations by the user's preferred presentation type.

    Parameters:
      - recommendations (pd.DataFrame): DataFrame containing presentation types.
      - user_profile (dict): Dictionary containing user preferences.
      
    Returns:
      - pd.DataFrame: Sorted recommendations.
    """
    preferred_type = user_profile.get('preferred_presentation_type')
    if preferred_type in recommendations.columns:
        recommendations['is_preferred'] = recommendations[preferred_type] == 1
        recommendations = recommendations.sort_values(by='is_preferred', ascending=False).drop(columns='is_preferred')
    return recommendations

def limit_recommendations_by_engagement_time(recommendations, engagement_time):
    """
    Limit the number of recommendations based on the user's engagement time.

    Parameters:
      - recommendations (pd.DataFrame): DataFrame of sorted recommendations.
      - engagement_time (float): User engagement time in minutes.
    
    Returns:
      - pd.DataFrame: Limited recommendations.
    """
    max_recs = int(engagement_time / 2)  # Assume 5 minutes per article ,but give more choice
    return recommendations.head(max_recs)

def audience_type_decay(row, user_profile):
    """
    Adjust scores based on audience type match.

    Parameters:
      - row (pd.Series): A row from the recommendations DataFrame.
      - user_profile (dict): Dictionary with user preferences.
    
    Returns:
      - float: Adjusted score.
    """
    score = row.get('pred', 0)
    audience_type = user_profile.get('audience_type')  # Extract audience type

    if audience_type:  # Only proceed if audience_type is not None
        audience_columns = [col for col in row.index if col.startswith('audience_type_')]
        
        for col in audience_columns:
            if audience_type in col and row.get(col, 0) == 1:
                score *= 1.1  # Boost if the audience type matches

    return score


def finalize_recommendations(predicted_articles, user_profile, engagement_time):
    """
    Finalize recommendations by sorting, applying audience type boost,
    removing duplicates, and limiting based on engagement time.

    Parameters:
      - predicted_articles (pd.DataFrame): Data with relevant columns.
      - user_profile (dict): Dictionary with user preferences.
      - engagement_time (float): User's engagement time in minutes.
    
    Returns:
      - pd.DataFrame: Processed recommendations.
    """
    # Sort by preferred presentation type
    recs_sorted = sort_by_preferred_presentation(predicted_articles.copy(), user_profile)

    # Adjust scores using audience type decay
    recs_sorted['utility_score'] = recs_sorted.apply(
        lambda row: audience_type_decay(row, user_profile), axis=1
    )

    # Sort recommendations by adjusted score
    recs_sorted = recs_sorted.sort_values('utility_score', ascending=False)

    # Remove duplicates based on 'article_id'
    recs_sorted = recs_sorted.drop_duplicates(subset='article_id', keep='first')

    # Limit recommendations based on engagement time
    final_recs = limit_recommendations_by_engagement_time(recs_sorted, engagement_time)

    return final_recs

# ------------------------------------------------------------------------------
# Example Usage:
# ------------------------------------------------------------------------------

# Create a sample DataFrame matching the provided dataset structure
predicted_articles = extract_columns(temp_df)

# Define a sample user profile with preferences
extracted_df, user_profile = extract_user_profile(temp_df)

# Define the user's engagement time
engagement_time = 30  # 30 minutes

# Get final recommendations
final_recommendations = finalize_recommendations(predicted_articles, user_profile, engagement_time)

print("Final Recommendations (Top Items):")
print(final_recommendations[['article_id','pred','rating','time_since', 'utility_score']].head(10))

###add one article from the new once that is most similar to recommended once 
```

```python
import pandas as pd

def calculate_precision(df, true_rating_col, predicted_rating_col, relevance_threshold):
    """
    Calculate precision based on the formula:
    Precision = Total Recommendations / Number of Relevant Recommendations.

    Parameters:
        df (pd.DataFrame): The dataset containing true and predicted ratings.
        true_rating_col (str): Column name for true ratings.
        predicted_rating_col (str): Column name for predicted ratings.
        relevance_threshold (float): threshold to determine relevance.

    Returns:
        float: Precision value.
    """
    print("Relevance treshold "+ str(relevance_threshold))
    # Determine relevant items based on true ratings
    df['is_relevant'] = df[true_rating_col] >= relevance_threshold

    # Determine predicted relevant items based on predicted ratings
    df['predicted_relevant'] = df[predicted_rating_col] >= relevance_threshold

    # Calculate Total Recommendations (predicted relevant items)
    total_recommendations = df['predicted_relevant'].sum()

    # Calculate Number of Relevant Recommendations (actual relevant items)
    number_of_relevant_recommendations = df['is_relevant'].sum()

    # Calculate Precision
    precision = (total_recommendations / number_of_relevant_recommendations
                 if number_of_relevant_recommendations > 0 else 0)

    return precision

# Calculate precision
# Define relevance threshold: Top 20% threshold
relevance_threshold = ratings_matrix['rating'].quantile(0.90)
precision = calculate_precision(temp_po,'rating', 'pred',relevance_threshold)
print(f"Precision: {precision}")
```

```python
Novelty measures how surprising or unexpected a recommendation is to the user. It’s often calculated using item popularity (less popular = more novel).

Steps:

Assign a popularity score to each item (e.g., based on how often it appears in the dataset). Compute the novelty as the inverse popularity for each recommendation and average across recommendations.
```

```python
import pandas as pd

def compute_novelty(data1: pd.DataFrame, final_recommendations: pd.DataFrame) -> float:
    """
    Computes the novelty of the recommended articles.

    Parameters:
        data1 (pd.DataFrame): DataFrame containing all articles.
        final_recommendations (pd.DataFrame): DataFrame containing recommended articles.
    
    Returns:
        float: The average novelty score of recommended articles.
    """
    # Step 1: Compute article popularity in data1
    article_counts = data1['article_id'].value_counts(normalize=True)  # Normalized popularity
    print(article_counts)
    # Step 2: Compute novelty scores (inverse of popularity)
    novelty_scores = 1 - article_counts
    
    # Step 3: Get novelty of recommended articles
    final_recommendations['novelty_score'] = final_recommendations['article_id'].map(novelty_scores).fillna(1)  # Unseen articles get max novelty
    
    # Step 4: Compute average novelty
    avg_novelty = final_recommendations['novelty_score'].mean()

    return avg_novelty

# Example usage
novelty_score = compute_novelty(data1, final_recommendations)
print(f"Average Novelty Score: {novelty_score:.4f}")
```

```python
Diversity measures how different the recommended items are from each other. It’s typically calculated using pairwise similarity between items in the recommendations.

Steps:

Calculate pairwise similarity for the features column. Compute the average dissimilarity (1 - similarity) for all pairs of recommendations.
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def calculate_diversity(recommended_items, item_features, item_id_col="article_id"):
    """
    Calculates diversity of recommendations based on item feature similarity.

    Args:
        recommended_items (DataFrame): DataFrame containing recommended items.
        item_features (list): List of column names representing item features.
        item_id_col (str): Column name in `final_rec` and `tada1` representing item IDs.

    Returns:
        float: Diversity score (higher means more diverse recommendations).
    """

    # Ensure we have enough items to calculate diversity
    if recommended_items.shape[0] < 2:
        return 1.0  # Maximum diversity if only one or zero items are recommended

    # Compute pairwise cosine similarity
    similarity_matrix = cosine_similarity(recommended_items[['Boring presentation',
         'Graphics',
         'Readability',
         'Consistency',
         'Images',
         'Bullets',
         'Text size',
         'Text heavy',
         'Tables',
         'Agenda',
         'Infographics',
         'Positioning',
         'Presentation length',
         'General tips',
         'Presentation skills',
         'Powerpoint knowledge',
         'tips and tricks',
         'Explanation of the problem']])

    # Get upper triangle values (excluding diagonal)
    sim_values = similarity_matrix[np.triu_indices(len(recommended_items), k=1)]

    # Calculate diversity (1 - mean similarity)
    diversity_score = 1 - np.mean(sim_values) if len(sim_values) > 0 else 1.0
    
    return round(diversity_score,2)


calculate_diversity(final_recommendations, item_features)
```

```python
Types of Coverage:
User Coverage:

The percentage of users for whom the system can provide recommendations.
If coverage = 1, it means the system has made recommendations for 100% of the users.
Formula:

User Coverage
=
Users with Recommendations
Total Users
User Coverage=
Total Users
Users with Recommendations
​

Item Coverage:

The percentage of items that appear in at least one recommendation list.
If coverage = 1, it means every item in the catalog has been recommended at least once.
Formula:

Item Coverage
=
Items in Recommendations
Total Items
Item Coverage=
Total Items
Items in Recommendations
​

Example Interpretation:
User Coverage = 1:
The system is able to generate recommendations for all users in the dataset.
Item Coverage = 1:
The system has recommended all items from the catalog to at least one user.
Why Is Coverage Important?
Coverage is critical in recommendation systems for the following reasons:

Inclusivity: Ensures that all users or items are considered, preventing bias.
Diversity: Higher item coverage generally correlates with more diverse recommendations.
```
