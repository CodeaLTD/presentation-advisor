```python
import random
import pandas as pd
from datetime import datetime, timedelta

def generate_presentations_for_users(num_users, problems, max_issues_per_presentation=6):
    """
    Generate presentations for a given number of users with random issues marked.

    Parameters:
        num_users (int): The number of users to generate presentations for.
        problems (list): List of possible problems to mark in presentations.
        max_issues_per_presentation (int): Maximum number of issues per presentation (default: 6).

    Returns:
        pd.DataFrame: A DataFrame containing all generated presentations.
    """
    def generate_random_date():
        """
        Generate a random date within the last two years.

        Returns:
            str: A string representing the random date in 'YYYY-MM-DD' format.
        """
        today = datetime.today()
        two_years_ago = today - timedelta(days=365*2)
        random_days = random.randint(0, (today - two_years_ago).days)
        random_date = two_years_ago + timedelta(days=random_days)
        return random_date.strftime("%Y-%m-%d")

    def generate_presentation(user_id, presentation_count):
        """
        Generate a random presentation with issues marked for a specific user.

        Parameters:
            user_id (int): The ID of the user.
            presentation_count (int): Counter to ensure unique presentation IDs.

        Returns:
            dict: A dictionary representing a presentation with marked issues.
        """
        # Initialize all problems with 0
        presentation = {problem: 0 for problem in problems}
        # Select a random number of problems to mark as 1
        num_problems = random.randint(1, max_issues_per_presentation)
        selected_problems = random.sample(problems, num_problems)
        for problem in selected_problems:
            presentation[problem] = 1
        presentation['user_id'] = user_id
        presentation['presentation_id'] = f"U{user_id}_P{presentation_count}"
        presentation['submission_date'] = generate_random_date()  # Add random submission date
        return presentation

    presentations = []
    presentation_count = 1  # Initialize a counter for unique presentation IDs
    for user_id in range(1, num_users + 1):
        # Each user gets at least one presentation
        presentations.append(generate_presentation(user_id, presentation_count))
        presentation_count += 1
        # Some users randomly get an additional presentation
        if random.choice([True, False]):
            presentations.append(generate_presentation(user_id, presentation_count))
            presentation_count += 1

    return pd.DataFrame(presentations)

```

```python
import pandas as pd
import random

def generate_article_ratings_with_preferences_and_popularity(
    user_profiles, recommendations, presentations_df, existing_ratings=None, min_rating=1, max_rating=5
):
    """
    Generate article ratings based on user-linked presentations, issues, audience type,
    preferred presentation type, and article popularity.

    Parameters:
        user_profiles (pd.DataFrame): DataFrame containing user profiles.
        recommendations (pd.DataFrame): DataFrame containing article recommendations.
        presentations_df (pd.DataFrame): DataFrame linking users to presentations with issues and submission dates.
        existing_ratings (pd.DataFrame, optional): DataFrame containing existing ratings for articles. Defaults to None.
        min_rating (int): Minimum possible rating value.
        max_rating (int): Maximum possible rating value.

    Returns:
        pd.DataFrame: A DataFrame containing user_id, article_id, and ratings.
    """
    ratings = []

    # Convert submission_date to datetime for prioritization
    presentations_df["submission_date"] = pd.to_datetime(presentations_df["submission_date"])

    # Calculate average article ratings if existing ratings are provided
    if existing_ratings is not None:
        article_avg_ratings = existing_ratings.groupby("article_id")["rating"].mean()
    else:
        article_avg_ratings = pd.Series(dtype=float)

    for _, user in user_profiles.iterrows():
        user_id = user["user_id"]
        user_preferences = user["preferences"]
        user_type = user["type"]
        preferred_audience = audience_mapping.get(user_type, "general")
        preferred_presentation_type = user["user_preferred_presentation_type"]

        # Filter presentations linked to this user
        user_presentations = presentations_df[presentations_df["user_id"] == user_id]

        # Weight presentations based on recency
        user_presentations["recency_weight"] = (
            1 / (1 + (pd.Timestamp.now() - user_presentations["submission_date"]).dt.days)
        )

        # Aggregate issues for this user's presentations, weighted by recency
        weighted_issues = (
            user_presentations.iloc[:, :-3]
            .multiply(user_presentations["recency_weight"], axis=0)
            .sum()
        )

        for _, rec in recommendations.iterrows():
            article_id = rec["article_id"]
            presentation_type = rec["presentation_type"]
            main_issue = rec["Main issue"]
            audience_type = rec["audience_type"]

            # Base rating
            rating = (max_rating + min_rating) // 2  # Start with midpoint rating

            # Rule 1: Match user preferences with weighted issues
            matching_issues = set(user_preferences) & set(weighted_issues.index[weighted_issues > 0])
            if matching_issues:
                rating += len(matching_issues)  # Increase rating for each matching issue

            # Rule 2: Match preferred presentation type
            if preferred_presentation_type == presentation_type:
                rating += 1  # Boost rating for matching presentation type
            else:
                rating -= 0.5  # Penalize for non-matching presentation type

            # Rule 3: Match audience type (skip if preferred audience type is "general")
            if preferred_audience != "general":
                if audience_type == preferred_audience:
                    rating += 0.5  # Boost rating for matching audience type
                else:
                    rating -= 0.5 # Penalize for non-matching audience type

            # Rule 4: Incorporate main issue
            if main_issue in user_preferences:
                rating += 0.5

            # Rule 5: Adjust based on article popularity (average rating)
            if article_id in article_avg_ratings:
                avg_rating = article_avg_ratings[article_id]
                if avg_rating > (max_rating + min_rating) / 2:
                    rating += 0.5  # Boost for higher average ratings
                else:
                    rating -= 0.5  # Penalize for lower average ratings

            # Ensure rating is within bounds
            rating = max(min_rating, min(rating, max_rating))

            # Append the rating
            ratings.append({"user_id": user_id, "article_id": article_id, "rating": rating})

    return pd.DataFrame(ratings)


import numpy as np
import pandas as pd

def generate_ratings_matrix(user_profiles_df, recommendations, presentations_df, existing_ratings=None, zero_percentage=0.2):
    """
    Generate a ratings matrix and replace a specified percentage of the ratings with 0.

    Parameters:
        user_profiles_df (pd.DataFrame): User profiles DataFrame.
        recommendations (pd.DataFrame): Recommendations DataFrame.
        presentations_df (pd.DataFrame): Presentations DataFrame.
        existing_ratings (pd.DataFrame, optional): Existing ratings DataFrame.
        zero_percentage (float): Percentage of ratings to replace with 0 (0 <= zero_percentage <= 1).

    Returns:
        pd.DataFrame: Ratings matrix with a percentage of ratings replaced with 0.
    """
    # Step 1: Generate initial ratings
    article_ratings = generate_article_ratings_with_preferences_and_popularity(
        user_profiles_df, recommendations, presentations_df, existing_ratings
    )

    # Remove duplicates based on 'user_id' and 'article_id'
    article_ratings = article_ratings.drop_duplicates(subset=['user_id', 'article_id'], keep='first')


    # Step 2: Replace x% of the ratings with 0
    total_rows = len(article_ratings)  # Total number of rows in the DataFrame
    num_to_replace = int(total_rows * zero_percentage)  # Number of rows to replace with 0

    # Randomly select indices to replace
    indices_to_replace = np.random.choice(article_ratings.index, size=num_to_replace, replace=False)

    # Replace selected rows' ratings with 0
    article_ratings_with_zeros = article_ratings.copy()
    article_ratings_with_zeros.loc[indices_to_replace, 'rating'] = 0

    return article_ratings, article_ratings_with_zeros



```

```python
def preprocess_user_profile(user_profile, preferences_options, types, presentation_type_options):
    """
    Preprocess user profile by flattening preferences and binary encoding categorical features.

    Parameters:
        user_profile (pd.DataFrame): DataFrame containing user-specific preferences and attributes.
        preferences_options (list): List of all possible preferences.
        types (list): List of possible user types.
        presentation_type_options (list): List of possible presentation types.

    Returns:
        pd.DataFrame: Updated user_profile with flattened preferences and binary-encoded features.
    """
    # Step 1: Flatten preferences
    preference_columns = {pref: f"user_preference_{pref}" for pref in preferences_options}
    preferences_df = pd.DataFrame(0, index=user_profile.index, columns=preference_columns.values())
    for i, prefs in enumerate(user_profile['preferences']):
        for pref in prefs:
            if pref in preference_columns:
                preferences_df.loc[i, preference_columns[pref]] = 1

    # Step 2: Binary encode 'type'
    type_columns = {t: f"user_type_{t}" for t in types}
    type_df = pd.DataFrame(0, index=user_profile.index, columns=type_columns.values())
    for i, user_type in enumerate(user_profile['type']):
        if user_type in type_columns:
            type_df.loc[i, type_columns[user_type]] = 1

    # Binary encode 'user_preferred_presentation_type'
    presentation_columns = {pt: f"user_presentation_type_{pt}" for pt in presentation_type_options}
    presentation_df = pd.DataFrame(0, index=user_profile.index, columns=presentation_columns.values())
    for i, presentation_type in enumerate(user_profile['user_preferred_presentation_type']):
        if presentation_type in presentation_columns:
            presentation_df.loc[i, presentation_columns[presentation_type]] = 1

    # Concatenate all features with the original DataFrame
    user_profile_encoded = pd.concat([
        user_profile.drop(columns=['preferences', 'type', 'user_preferred_presentation_type']),
        preferences_df, type_df, presentation_df
    ], axis=1)

    return user_profile_encoded

# # Example user_profile
# user_profile = pd.DataFrame({
#     'user_id': [1, 2, 3],
#     'type': ['business', 'specialist', 'teacher'],
#     'location': ['New York', 'Florida', 'California'],
#     'preferences': [
#         ['Presentation skills', 'Bullets'],
#         ['General tips'],
#         ['Agenda', 'Text heavy', 'Consistency']
#     ],
#     'preferred_presentation_type': ['formal', 'creative', 'creative']
# })

# preferences_options = [
#     "Boring presentation", "Graphics", "Readability", "Consistency",
#     "Images", "Bullets", "Text size", "Text heavy", "Tables", "Agenda",
#     "Infographics", "Positioning", "Presentation length",
#     "General tips", "Presentation skills", "Powerpoint knowledge",
#     "tips and tricks", "Explanation of the problem"
# ]

# types = ['business', 'teacher', 'student', 'researcher', 'manager', 'technical', 'specialist']
# presentation_type_options = ["formal", "creative", "students", "technical", "branding"]

# # Preprocess user profile
# user_profile_encoded = preprocess_user_profile(user_profile, preferences_options, types, presentation_type_options)

# # Display the updated user profile
# print(user_profile_encoded)

```

```python
def is_all_categorical(df):
    """
    Checks if all columns in the DataFrame are categorical.

    Parameters:
    - df: DataFrame to check.

    Returns:
    - bool: True if all columns are categorical, False otherwise.
    """
    # Check if all columns are of type object or category
    all_categorical = all(df.dtypes.apply(lambda dtype: dtype.name in ['object', 'category']))
    return all_categorical

def categorize_columns(df):
    """
    Categorizes columns into categorical and non-categorical.

    Parameters:
    - df: DataFrame to analyze.

    Returns:
    - dict: A dictionary with lists of categorical and non-categorical columns.
    """
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    non_categorical_cols = df.select_dtypes(exclude=['object', 'category']).columns.tolist()
    return {'categorical': categorical_cols, 'non_categorical': non_categorical_cols}

def analyze_categorical_columns(df):
    """
    Analyzes columns in a DataFrame to determine:
    - Columns that are binary (0/1).
    - Columns that can be one-hot encoded (more than 2 unique values).

    Parameters:
    - df: DataFrame to analyze.

    Returns:
    - dict: A dictionary with:
        - 'binary_columns': List of columns with only 0 and 1 values.
        - 'one_hot_columns': List of columns that can be one-hot encoded.
    """
    binary_columns = []
    one_hot_columns = []

    for col in df.columns:
        unique_values = df[col].unique()
        if set(unique_values) == {0, 1} or set(unique_values) == {0} or set(unique_values) == {1}:
            binary_columns.append(col)
        elif df[col].dtype == 'object' or len(unique_values) > 2:
            one_hot_columns.append(col)

    return {
        'binary_columns': binary_columns,
        'one_hot_columns': one_hot_columns
    }

```

Precision= Total recommendations/Number of relevant recommendations


```python
import pandas as pd

def calculate_precision(df, true_rating_col, predicted_rating_col, relevance_threshold):
    """
    Calculate precision based on the formula:
    Precision = Total Recommendations / Number of Relevant Recommendations.

    Parameters:
        df (pd.DataFrame): The dataset containing true and predicted ratings.
        true_rating_col (str): Column name for true ratings.
        predicted_rating_col (str): Column name for predicted ratings.
        relevance_threshold (float): threshold to determine relevance.

    Returns:
        float: Precision value.
    """

    # Determine relevant items based on true ratings
    df['is_relevant'] = df[true_rating_col] >= relevance_threshold

    # Determine predicted relevant items based on predicted ratings
    df['predicted_relevant'] = df[predicted_rating_col] >= relevance_threshold

    # Calculate Total Recommendations (predicted relevant items)
    total_recommendations = df['predicted_relevant'].sum()

    # Calculate Number of Relevant Recommendations (actual relevant items)
    number_of_relevant_recommendations = df['is_relevant'].sum()

    # Calculate Precision
    precision = (total_recommendations / number_of_relevant_recommendations
                 if number_of_relevant_recommendations > 0 else 0)

    return precision

# Calculate precision
# Define relevance threshold: Top 20% threshold
relevance_threshold = ratings_matrix['rating'].quantile(0.80)
precision = calculate_precision(hybrid_recommendations,'true_rating', 'predicted_rating',relevance_threshold)
print(f"Precision: {precision}")
```


Novelty measures how surprising or unexpected a recommendation is to the user. It’s often calculated using item popularity (less popular = more novel).

Steps:

Assign a popularity score to each item (e.g., based on how often it appears in the dataset).
Compute the novelty as the inverse popularity for each recommendation and average across recommendations.

```python
import pandas as pd

def calculate_novelty_by_inverse_popularity(df, item_col):
    """
    Calculate novelty as the average inverse of item popularity.

    Parameters:
        df (pd.DataFrame): Dataset containing items for which novelty is to be calculated.
        item_col (str): Column name for the item IDs (e.g., 'article_id').

    Returns:
        float: Average novelty score.
    """
    # Calculate item popularity
    item_popularity = df[item_col].value_counts()

    # Assign novelty as the inverse of popularity
    df['novelty'] = 1 / df[item_col].map(item_popularity)

    # Calculate average novelty
    novelty_score = df['novelty'].mean()

    return novelty_score

novelty_score = calculate_novelty_by_inverse_popularity(hybrid_recommendations, item_col='article_id')
print(f"Novelty Score: {novelty_score}")
```

Diversity measures how different the recommended items are from each other. It’s typically calculated using pairwise similarity between items in the recommendations.

Steps:

Calculate pairwise similarity for the features column.
Compute the average dissimilarity (1 - similarity) for all pairs of recommendations.

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def calculate_diversity(df, item_col, feature_matrix):
    """
    Calculate diversity of recommendations as the average pairwise dissimilarity.

    Parameters:
        df (pd.DataFrame): Dataset containing recommended items.
        item_col (str): Column name for the item IDs (e.g., 'article_id').
        feature_matrix (pd.DataFrame or np.ndarray): Feature matrix of items (rows: items, cols: features).

    Returns:
        float: Diversity score (higher means more diverse recommendations).
    """
    # Extract recommended items
    recommended_items = df[item_col].unique()

    # Subset feature matrix for recommended items
    feature_subset = feature_matrix.loc[recommended_items]

    # Calculate pairwise cosine similarity
    similarity_matrix = cosine_similarity(feature_subset)

    # Convert similarity to dissimilarity (1 - similarity)
    dissimilarity_matrix = 1 - similarity_matrix

    # Calculate average pairwise dissimilarity
    num_items = len(recommended_items)
    if num_items < 2:
        return 0  # No diversity if fewer than 2 items are recommended

    # Take the upper triangle of the matrix (excluding diagonal)
    dissimilarity_values = dissimilarity_matrix[np.triu_indices(num_items, k=1)]
    diversity_score = dissimilarity_values.mean()

    return diversity_score
```

Types of Coverage:
User Coverage:

The percentage of users for whom the system can provide recommendations.
If coverage = 1, it means the system has made recommendations for 100% of the users.
Formula:

User Coverage
=
Users with Recommendations
Total Users
User Coverage=
Total Users
Users with Recommendations
​

Item Coverage:

The percentage of items that appear in at least one recommendation list.
If coverage = 1, it means every item in the catalog has been recommended at least once.
Formula:

Item Coverage
=
Items in Recommendations
Total Items
Item Coverage=
Total Items
Items in Recommendations
​

Example Interpretation:
User Coverage = 1:
The system is able to generate recommendations for all users in the dataset.
Item Coverage = 1:
The system has recommended all items from the catalog to at least one user.
Why Is Coverage Important?
Coverage is critical in recommendation systems for the following reasons:

Inclusivity: Ensures that all users or items are considered, preventing bias.
Diversity: Higher item coverage generally correlates with more diverse recommendations.


```python
import pandas as pd

def calculate_coverage(recommendations, user_col, item_col, total_users, total_items):
    """
    Calculate user and item coverage for a recommendation system.

    Parameters:
        recommendations (pd.DataFrame): DataFrame containing recommendations.
                                        Must include user and item columns.
        user_col (str): Column name for user IDs in the recommendations DataFrame.
        item_col (str): Column name for item IDs in the recommendations DataFrame.
        total_users (int): Total number of users in the dataset.
        total_items (int): Total number of items in the catalog.

    Returns:
        dict: A dictionary containing user coverage and item coverage.
    """
    # Users with recommendations
    users_with_recommendations = recommendations[user_col].nunique()

    # Items in recommendations
    items_in_recommendations = recommendations[item_col].nunique()

    # Calculate user coverage
    user_coverage = users_with_recommendations / total_users

    # Calculate item coverage
    item_coverage = items_in_recommendations / total_items

    return {
        'user_coverage': user_coverage,
        'item_coverage': item_coverage
    }
```

```python
# Example recommendation DataFrame
recommendations = hybrid_recommendations.copy()

# Total users and items in the dataset
total_users = 30  # Example: 6 users in the dataset
total_items = 300  # Example: 10 items in the catalog

# Calculate coverage
coverage = calculate_coverage(
    recommendations,
    user_col='user_id',
    item_col='article_id',
    total_users=total_users,
    total_items=total_items
)

print(f"User Coverage: {coverage['user_coverage']}")
print(f"Item Coverage: {coverage['item_coverage']}")
```
